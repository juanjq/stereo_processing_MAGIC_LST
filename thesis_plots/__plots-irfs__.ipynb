```python
gamma_array = np.linspace(0, 1, 200)

time_diffs = np.diff(df_mean['timestamp'].to_numpy())
obs_time   = time_diffs[time_diffs < 1].sum()
print(f'The total observation time is: {obs_time/3600:.2f} h')

print(f'\nGlobal theta2 cut: {theta2_cut} [deg2]\n')

# Calculate the numbers ON, OFF and excess events
dfs      = [df_mean.copy(), df_lst.copy(), df_magic.copy()]
dfs      = [df[['gammaness', 'theta2_on', 'reco_energy', *[f'theta2_off{i+1}' for i in range(n_regions_off)]]] for df in dfs]
label_df = ['3Tel-stereo', 'LST-lstchain', 'MAGIC-mars']
cols     = ['darkblue', 'crimson', 'c']

significance = [[] for _ in dfs]
prop_gamma   = [[] for _ in dfs]
gamma_rate   = [[] for _ in dfs]
off_rate     = [[] for _ in dfs]
n_ons        = [[] for _ in dfs]
n_offs       = [[] for _ in dfs]

tmp_dfs = dfs.copy()

for g in range(len(gamma_array)):
    
    if g % 20 == 0:
        print(f'Computing {g / len(gamma_array) * 100:.0f}%...')
        
    for i in range(len(dfs)):
        
        tmp_dfs[i] = tmp_dfs[i].query(f'gammaness >= {gamma_array[g]}', inplace=False)

        n_on  = sum(tmp_dfs[i]['theta2_on' ] < theta2_cut)
        n_off = sum([sum(tmp_dfs[i][f'theta2_off{j+1}'] < theta2_cut) for j in range(n_regions_off)])
        n_offs[i].append(n_off)
        n_ons[i].append(n_on)
        
        significance[i].append(li_ma_significance(n_on, n_off, alpha=1/n_regions_off))
        
        gamma_rate[i].append((n_on-n_off)/obs_time/60)
        off_rate[i].append(n_off/obs_time/60)
        
        if n_off !=0:
            prop_gamma[i].append(n_on/n_off)
        else:
            prop_gamma[i].append(0)
            
# calculate the optimal gammas
def neg_WStat(gammaness, i):
    
    df = dfs[i].query(f'gammaness >= {gammaness}', inplace=False)[['gammaness', 'theta2_on', *[f'theta2_off{i+1}' for i in range(n_regions_off)]]]
    
    n_on  = sum(df['theta2_on'] < theta2_cut)
    n_off = sum([sum(df[f'theta2_off{j+1}'] < theta2_cut) for j in range(n_regions_off)])
        
    return -1 * li_ma_significance(n_on, n_off, alpha=1/n_regions_off)

opt_gamma = []
for i in range(len(significance)):
    opt_gamma.append(minimize(neg_WStat, x0=[0.7], bounds=[[0,1]], args=(i), method='Nelder-Mead').x[0])
    
print(f'The maxiumum significances are given by:')
for i in range(len(significance)):
    maxgamma, maxsig = gamma_array[significance[i].index(max(significance[i]))], max(significance[i])
    nevs  = len(dfs[i][dfs[i]['gammaness'] >= maxgamma])
    nevsp = nevs / len(dfs[i]) * 100
    print(f'--> {label_df[i]}\t gamma = {opt_gamma[i]:.3f},  sigma = {maxsig:.1f},  nevs = {nevs}, {nevsp:.1f}%')
```

```python
# finallly we set up the optimal gamma to the datasets 

# we can apply this cuts to the datasets so
df_mean_F  =  df_mean.query( f'gammaness >= {opt_gamma[0]}')
df_lst_F   =  df_lst.query(  f'gammaness >= {opt_gamma[1]}')
df_magic_F =  df_magic.query(f'gammaness >= {opt_gamma[2]}')

# thetas of each dataframe filtered in gammaness
Ton_mean_F  = df_mean_F[ f'theta2_on'].to_numpy()
Ton_lst_F   = df_lst_F[  f'theta2_on'].to_numpy()
Ton_magic_F = df_magic_F[f'theta2_on'].to_numpy()

Toff_mean_F, Toff_lst_F, Toff_magic_F = [], [], []
for n in range(n_regions_off):
    Toff_mean_F.append( df_mean_F[ f'theta2_off{n+1}'].to_numpy())
    Toff_lst_F.append(  df_lst_F[  f'theta2_off{n+1}'].to_numpy())
    Toff_magic_F.append(df_magic_F[f'theta2_off{n+1}'].to_numpy())
```
