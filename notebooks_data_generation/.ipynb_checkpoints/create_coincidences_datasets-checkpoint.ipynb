Import packages and setting parameters

```python
# packages
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import yaml, os, logging, sys
pd.set_option("display.max_columns", None)

from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

# --- logging --- #
import logging
logger = logging.getLogger(__name__)
logger.addHandler(logging.StreamHandler())
logger.setLevel(logging.INFO)

# location of the scripts
sys.path.insert(0, '/fefs/aswg/workspace/juan.jimenez/stereo_analysis/scripts')
import auxiliar as aux
aux.params()

# --- other parameters --- #
# name of the source we are studying
source_name = 'BLLac'
# epoch we want to study {'all', '1', '2'}
epoch = 'all'
# ------------------------ #

# --- file paths --- #
# common data file
common_data_path = f'/fefs/aswg/workspace/juan.jimenez/stereo_analysis/config_files/common_data{source_name}.txt'
# output path to coincidences dataframes
path_coincidences = '/fefs/aswg/workspace/juan.jimenez/data/dl1/other_data/'
# configuration file
config_file = f'/fefs/aswg/workspace/juan.jimenez/stereo_analysis/config_files/config{source_name}.yaml'
# ------------------ #

logger.info(f'Study of the source: {source_name}\nIn the selected epoch: \'{epoch}\'')
```

Reading `common_data` file and finding the jobs. Preparing a job that contains all lst files.

```python
# first of all we can read the common data file
logger.info(f'Opening the file...\n{common_data_path}\n')

# saving the same-night runs comparing LST runs with all MAGIC
jobs_list = np.loadtxt(common_data_path, dtype='str')

logger.info(f'Found {len(jobs_list)} jobs:\n')

# extracting all LST runs
LST_runs, MAGIC_runs = [], []
for job in jobs_list:
    midindex = job.find('-')
    LST_runs.append(int(job[:midindex]))
    MAGIC_runs = MAGIC_runs + list(np.array(job[midindex+1:].split(',')).astype(int))
    
# for the magic runs we can smply select a short night
MAGIC_runs = ['5093174','5093175'] #np.unique(np.array(MAGIC_runs)).astype(str)

LST_runs   = np.array(LST_runs).astype(str)

# creating a unique total job
LST_str, MAGIC_str = ','.join(LST_runs), ','.join(MAGIC_runs)
obs_id_string = f'{LST_str}-{MAGIC_str}'

logger.info(f'The full job is constructed as:\n{obs_id_string}')

import find_files

# simply separating the MAGIC and LST runs
obs_id_LST, obs_id_MAGIC = find_files.str_to_run_ids(obs_id_string)
logger.info(f'\nSelected runs:\nLST runs   ---> {obs_id_LST}')
logger.info(f'MAGIC runs ---> {obs_id_MAGIC}')

# separating diferent epochs
if   epoch == '1':
    obs_id_LST = [2758, 2759, 2760, 2761, 2918, 2919, 2923, 2924, 2960, 2961, 2976, 2977, 3093, 3094, 3095, 3096, 3265, 3266, 3267, 3268,
                  3382, 3383, 3385, 3631, 3632, 3633, 3634, 3635, 3636, 3953, 3954, 3955, 3956, 4007, 4008, 4009, 4010, 4011, 4012, 4013,
                  4014, 4066, 4067, 4068, 4069, 4070, 4071, 4073, 4074, 4075, 4125, 4126, 4127, 4128, 4129]
elif epoch == '2':
    obs_id_LST = [5738, 5759, 6968, 6969, 7171, 7172, 7197, 7198, 7200, 7201, 7227, 7228, 7229, 7230, 7231, 7232, 7233, 7234, 7235, 7236,
                  7237, 7238, 7255, 7256, 7258, 7277, 7278, 7279, 7280, 7281, 7282, 7305, 7306]
```

Finding the calibrated MAGIC files

```python
# finding M1 and M2 calibrated files
files_M1_cal, files_M2_cal = find_files.find_MAGIC_cal(obs_id_MAGIC)

# finding dl1 directories for LST
files_LST_dl1 = find_files.find_LST_dl1(obs_id_LST)

# filtering the empty ones to all data arrays
obs_id_MAGIC = [obsid for obsid, i in zip(obs_id_MAGIC, range(len(files_M1_cal))) if len(files_M1_cal[i]) > 0]        
files_M1_cal = [f     for f,     i in zip(files_M1_cal, range(len(files_M1_cal))) if len(files_M1_cal[i]) > 0] 
files_M2_cal = [f     for f,     i in zip(files_M2_cal, range(len(files_M2_cal))) if len(files_M2_cal[i]) > 0] 

# creating again the obs_id_string
joint_MAGIC_ids =  ','.join(np.array(obs_id_MAGIC).astype(str))
obs_id_string   = f'{obs_id_LST[0]}-{joint_MAGIC_ids}'

# opening configuration
with open(config_file, 'rb') as f:
    config = yaml.safe_load(f) 
```

Now skipping the different analysis steps. Only reading the data paths.

```python
import magicruns_cal_to_dl1

# main output directory
output_dir = '/fefs/aswg/workspace/juan.jimenez/data/dl1/runs_magic_m1_and_m2'
process    = False # false if files aready processed to dl1 set to False

files_M1_dl1 = magicruns_cal_to_dl1.convert(files_M1_cal, obs_id_MAGIC, output_dir, config, 1, process)
files_M2_dl1 = magicruns_cal_to_dl1.convert(files_M2_cal, obs_id_MAGIC, output_dir, config, 2, process)  

import merge

input_dir  = '/fefs/aswg/workspace/juan.jimenez/data/dl1/runs_magic_m1_and_m2'
output_dir = '/fefs/aswg/workspace/juan.jimenez/data/dl1/nights_magic'
process    = False # false if files aready merged

files_MAGIC_dl1 = merge.MAGIC_dl1(obs_id_MAGIC, files_M1_dl1, files_M2_dl1, input_dir, output_dir, process)

import event_coincidence_night

output_dir = '/fefs/aswg/workspace/juan.jimenez/data/dl1/joint_runs_magic_lst'
process    = False # if the coincident events already done

files_coinc_dl1 = event_coincidence_night.coincidences(obs_id_LST, files_LST_dl1, files_MAGIC_dl1, output_dir, config, process, process_windows=False)
```

Not all files exist (by coincidence of events and validity of some subruns) so we will save everything that exists in one file of each type

```python
df_feature = []
df_profile = []
for run in files_coinc_dl1:
    for subrun in run:      
        try:
            df_profile.append(pd.read_hdf(subrun, key='coincidence/profile'))
            df_feature.append(pd.read_hdf(subrun, key='coincidence/feature'))
        except: 
            pass
        
PROFILE = df_profile[0]
for i in range(len(df_profile)-1):
    for col in ['n_coincidence_tel12', 'n_coincidence_tel13']:
        PROFILE[col] = PROFILE[col] + df_profile[i+1][col]
    
FEATURE = pd.concat(df_feature)

logger.info(f'Profile data:')
display(PROFILE.head(5))
logger.info(f'Feature data:')
display(FEATURE.head(10))
```

Writting and saving the dataframes

```python
# saving the dataframes
if epoch == 'all':
    sub_name = 'total'
elif epoch == '1':
    sub_name = 'first_epoch'
elif epoch == '2':
    sub_name = 'second_epoch'
else:
    logger.error(f'Given epoch "{epoch}" is not valid, needed one of: ("all", "1", "2")')
    
FEATURE.to_csv(os.path.join(path_coincidences, f'feature_data_{source_name}.{sub_name}.csv'), index=False)
PROFILE.to_csv(os.path.join(path_coincidences, f'profile_data_{source_name}.{sub_name}.csv'), index=False)
```
