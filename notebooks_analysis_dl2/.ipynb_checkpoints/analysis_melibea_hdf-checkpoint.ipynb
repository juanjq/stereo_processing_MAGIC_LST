# Reading MELIBEA data

```python
# root directory for melibea h5 data
root = '/fefs/aswg/workspace/juan.jimenez/data/dl2/melibea'

# mask name of the melibea files
name = 'melibeaCommonCrabNebula.*' # important to put the final point

# Find the input files
input_files = glob.glob(os.path.join(root,name))
input_files.sort()

# reading dataframes and putting all together into a unique dataframe
melibea_dfs = [pd.read_hdf(f) for f in input_files]
melibea_df = pd.concat(melibea_dfs)

print(f'\nNumber of MELIBEA events: {len(melibea_df)}\n')
melibea_df
```

# Extracting data to arrays

```python
df = melibea_df

# energy and gammaness
en, gamma            = df['reco_energy'].to_numpy()     , df['gammaness'].to_numpy()
# MAGIC 1 and 2 intensities
inten_1, inten_2     = df['intensity_M1'].to_numpy(), df['intensity_M2'].to_numpy()
# # X, Y direction
dirX, dirY           = df['st_dirX'].to_numpy()     , df['st_dirY'].to_numpy()
# X, Y direction disp
dirX_disp, dirY_disp = df['st_dirX_disp'].to_numpy(), df['st_dirY_disp'].to_numpy()
# decayment and right ascension
dec, ra              = df['st_dec'].to_numpy()      , df['st_ra'].to_numpy()

# energy with gammaness filtering
en_f1   = (df.drop(df[df['gammaness'] < 0.1].index ))['reco_en'].to_numpy()
en_f5   = (df.drop(df[df['gammaness'] < 0.5].index ))['reco_en'].to_numpy()
en_f7   = (df.drop(df[df['gammaness'] < 0.7].index ))['reco_en'].to_numpy()                                                                        
en_f9   = (df.drop(df[df['gammaness'] < 0.9].index ))['reco_en'].to_numpy()
en_f99  = (df.drop(df[df['gammaness'] < 0.99].index))['reco_en'].to_numpy()                                                                         
```

# Reconstructed direction

```python
# direction reconstruction in degrees 

fig, ax = plt.subplots(figsize=(8,8))

hist = ax.hist2d(dirX_disp, dirY_disp, bins=100)
fig.colorbar(hist[3], ax=ax, label='counts')

ax.set_title(f'Recons. direction for {len(en)} crab ev.')
ax.set_xlabel('x_disp')
ax.set_ylabel('y_disp')
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
from astropy import units as u
import astropy.coordinates as coord
from astropy.coordinates import SkyCoord, Distance, Galactic, EarthLocation, AltAz
import matplotlib as mpl

c =  SkyCoord(ra=ra*u.deg, dec=dec*u.deg, frame='icrs')
c = c.galactic
```

# Energy reconstruction

```python
bins = np.logspace(0.3,6.05,120)

fig, ax = plt.subplots(figsize=(12,6))


ax.hist(en, bins=bins, label=f'Energy all ev', color='royalblue', histtype='stepfilled')
ax.hist(en_f1, bins=bins, label=f'E $\\gamma > 0.1$',color='crimson', histtype='stepfilled')
ax.hist(en_f5, bins=bins, label=f'E $\\gamma > 0.5$',color='c', histtype='stepfilled')
ax.hist(en_f7, bins=bins, label=f'E $\\gamma > 0.7$',color='gold', histtype='stepfilled')
ax.hist(en_f9, bins=bins, label=f'E $\\gamma > 0.99$',color='deeppink', histtype='stepfilled')


ax.loglog()
ax.legend(loc=1)
ax.set_title(f'Energy reconstruction for {len(en)} crab events')
ax.set_xlabel('Energy (GeV)')
ax.set_ylabel('counts')
plt.show()
```

# Intenisities

```python
bins = np.logspace(1.4,5.1,100)

fig, ax = plt.subplots(figsize=(12,6))


ax.hist(inten_1, bins=bins, label='Intensity MAGIC 1', histtype='stepfilled',alpha=0.8, color='mediumblue')
ax.hist(inten_2, bins=bins, histtype='stepfilled',alpha=0.4, lw=3, label='Intensity MAGIC 2', color='c')

ax.loglog()
ax.legend()
ax.set_title(f'intenisties for {len(en)} crab events')
ax.set_xlabel('Intensity (size)')
ax.set_ylabel('counts')
plt.show()
```

# Gammaness distribution

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.hist(gamma, bins=150, label='Gammaness melibea', color='royalblue')

ax.set_yscale('log')
ax.legend()
ax.set_title(f'intenisties for {len(en)} crab events')
ax.set_xlabel('gammaness')
ax.set_ylabel('counts')
plt.show()
```
