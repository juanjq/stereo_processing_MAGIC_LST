## 3D plot

```python
def z_projection(x0, y0, z0, zd, az, altitude):
    
    zd, az = np.deg2rad(zd), np.deg2rad(az)

#     if zd != 0:
    t = (altitude - z0) / np.cos(zd)
    x = x0 + t * np.cos(az) * np.sin(zd)
    y = y0 + t * np.sin(az) * np.sin(zd)
    z = altitude
#     else:
#         x = x0
#         y = y0
#         z = altitude
    
    return x, y, z

def z_plane(x, y, point, direction):
    
    x0, y0, z0 = point
    zd, az = direction
    zd, az = np.deg2rad(zd), np.deg2rad(az)
    
    ux, uy, uz = np.cos(az) * np.sin(zd), np.sin(az) * np.sin(zd), np.cos(zd)
    
    xy = (x - x0) * ux + (y - y0) * uy
    
    return z0 - xy / uz


def plane_projection(plane_point, plane_direction, xl, yl, zl, zdl, azl):
    
    xp, yp, zp = plane_point
    zdp, azp   = plane_direction
    
    # converting to rad
    zdp, azp = np.deg2rad(zdp), np.deg2rad(azp)
    zdl, azl = np.deg2rad(zdl), np.deg2rad(azl)
    
    # director vectors
    upx, upy, upz = np.cos(azp) * np.sin(zdp), np.sin(azp) * np.sin(zdp), np.cos(zdp)
    ulx, uly, ulz = np.cos(azl) * np.sin(zdl), np.sin(azl) * np.sin(zdl), np.cos(zdl)
    ulup = upx * ulx + upy * uly + upz * ulz
    
    t = ((xp - xl) * upx + (yp - yl) * upy + (zp - zl) * upz) / ulup
    
    x = xl + t * ulx
    y = yl + t * uly
    z = zl + t * ulz
    
    return x, y, z
```

## Projection test

```python
point_zd, point_az = 25, 45
point_dir = (point_zd, point_az)
print(f'The pointing direction is zd = {point_zd:.2f}deg and az = {point_az:.2f}deg')

N = 300

zd = np.random.normal(point_zd, 7, N)
az = np.random.normal(point_az, 7, N)

mean_h = 10
h_max  = np.random.normal(mean_h, 3, N)

core_x = np.random.normal(0, 5, N)
core_y = np.random.normal(0, 5, N)
core_z = np.zeros(N)


# --- reference points --- #
ref_core   = [0, 0, 0]
ref_source = z_projection(*ref_core, *point_dir, np.mean(h_max))
print(f'\nRef point for cores is  [{ref_core[0]:.2f}, {ref_core[1]:.2f}, {ref_core[2]:.2f}]m')
print(f'Ref point for source is [{ref_source[0]:.2f}, {ref_source[1]:.2f}, {ref_source[2]:.2f}]m')

MATRIX = {'reco_zd':zd, 'reco_az':az, 'h_max':h_max, 'core_x':core_x, 'core_y':core_y, 'core_z':core_z}

data = pd.DataFrame(MATRIX)

# --- calculating plane projections --- #
data['perp_x'], data['perp_y'], data['perp_z'] = plane_projection(ref_core, point_dir,
            data['core_x'], data['core_y'], data['core_z'], data['reco_zd'], data['reco_az'])
perp_x, perp_y, perp_z = data['perp_x'], data['perp_y'], data['perp_z']

data['proj_x'], data['proj_y'], data['proj_z'] = plane_projection(ref_core, point_dir,
            data['core_x'], data['core_y'], data['core_z'], point_zd, point_az)
proj_x, proj_y, proj_z = data['proj_x'], data['proj_y'], data['proj_z']

data['source_x'], data['source_y'], data['source_z'] = plane_projection(ref_source, point_dir,
            data['core_x'], data['core_y'], data['core_z'], data['reco_zd'], data['reco_az'])
source_x, source_y, source_z = data['source_x'], data['source_y'], data['source_z']

data['hsource_x'], data['hsource_y'], data['hsource_z'] = z_projection(
    data['core_x'], data['core_y'], data['core_z'], data['reco_zd'], data['reco_az'], data['h_max'])
hsource_x, hsource_y, hsource_z = data['hsource_x'], data['hsource_y'], data['hsource_z']

data
```

```python
%matplotlib notebook
%matplotlib inline

fig = plt.figure(figsize=(20,20))
ax  = fig.add_subplot(1, 2, 1, projection='3d') 



# --- plot the reference points --- #
ax.plot([ref_core[0],ref_source[0]], [ref_core[1], ref_source[1]], [ref_core[2], ref_source[2]],
        '--', c='k', marker='o', ms=6, lw=2, zorder=15)


# --- plotting limits --- #
delta_x = abs(ref_core[0] - ref_source[0])
delta_y = abs(ref_core[1] - ref_source[1])
delta_z = abs(ref_core[2] - ref_source[2])
xrefs, yrefs, zrefs = [ref_core[0], ref_source[0]], [ref_core[1], ref_source[1]], [ref_core[1], ref_source[1]]
factor = 2
xmin, xmax = min(xrefs) - delta_x * factor, max(xrefs) + delta_x * factor
ymin, ymax = min(yrefs) - delta_y * factor, max(yrefs) + delta_y * factor
zmin, zmax = min(zrefs) - delta_z * factor, max(zrefs) + delta_z * factor

# --- masks for the points --- #
core_mask = (core_x >= xmin) & (core_x <= xmax) & (core_y >= ymin) & (core_y <= ymax) & \
            (core_z >= zmin) & (core_z <= zmax)
perp_mask = (perp_x >= xmin) & (perp_x <= xmax) & (perp_y >= ymin) & (perp_y <= ymax) & \
            (perp_z >= zmin) & (perp_z <= zmax)
proj_mask = (proj_x >= xmin) & (proj_x <= xmax) & (proj_y >= ymin) & (proj_y <= ymax) & \
            (proj_z >= zmin) & (proj_z <= zmax)
source_mask = (source_x >= xmin) & (source_x <= xmax) & (source_y >= ymin) & (source_y <= ymax) & \
            (source_z >= zmin) & (source_z <= zmax)
hsource_mask = (hsource_x >= xmin) & (hsource_x <= xmax) & (hsource_y >= ymin) & (hsource_y <= ymax) & \
            (hsource_z >= zmin) & (hsource_z <= zmax)

# --- create the planes --- #
# x & y points for the planes
n_points_plane = 2
x_range = np.linspace(xmin, xmax, n_points_plane)
y_range = np.linspace(ymin, ymax, n_points_plane)
X_grid,  Y_grid  = np.meshgrid(x_range, y_range)
# reference planes
core_plane   = z_plane(X_grid, Y_grid, ref_core,   [0, 0])
perp_plane   = z_plane(X_grid, Y_grid, ref_core,   point_dir)
source_plane = z_plane(X_grid, Y_grid, ref_source, point_dir)


# --- plot the planes --- #
ax.plot_surface(X_grid, Y_grid, core_plane, alpha=0.6, color='gray', zorder=-9)
ax.plot_surface(X_grid, Y_grid, perp_plane, alpha=0.6, color='orangered', zorder=-10)
ax.plot_surface(X_grid, Y_grid, source_plane, alpha=0.6, color='royalblue', zorder=-8)

ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_zlabel('z [m]')

# setting the limits
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)


# --- random sample plotting --- #
ax.plot(core_x[core_mask], core_y[core_mask], core_z[core_mask], 'o', color='gray', zorder=11, label='core positions')
ax.plot(perp_x[perp_mask], perp_y[perp_mask], perp_z[perp_mask], 'o', color='r', zorder=9, label='$\perp$-plane trajectory ext.')
ax.plot(proj_x[proj_mask], proj_y[proj_mask], proj_z[proj_mask], 'o', color='maroon', zorder=10, ms=2, label='$\perp$-plane projection')
ax.plot(source_x[source_mask], source_y[source_mask], source_z[source_mask], 'o', color='b', zorder=19, label='h_max plane trajectory ext.')
ax.plot(hsource_x[hsource_mask], hsource_y[hsource_mask], hsource_z[hsource_mask], '.', color='c', zorder=19, label='h_max positions')

ax.legend(loc=9)
fig.tight_layout()
ax.view_init(4, 30)
ax.set_box_aspect([1,1,1])
plt.show()
```

```python
%matplotlib notebook
%matplotlib inline

fig = plt.figure(figsize=(20,10))
ax  = fig.add_subplot(1, 2, 1, projection='3d') 
ax1 = fig.add_subplot(2, 4, 3) 
ax2 = fig.add_subplot(2, 4, 4) 
ax3 = fig.add_subplot(2, 4, 7) 
ax4 = fig.add_subplot(2, 4, 8) 


# --- plot the reference points --- #
ax.plot([ref_core[0],ref_source[0]], [ref_core[1], ref_source[1]], [ref_core[2], ref_source[2]],
        '--', c='k', marker='', ms=6, lw=2, zorder=15)
ax.plot([*ypos, ypos[0]], [*xpos, xpos[0]], [*zpos, zpos[0]], 
        ls='', c='w', marker='+', zorder=20, ms=4, lw=2, alpha=0.6)
ax.plot([*xpos_upper, xpos_upper[0]], [*ypos_upper, ypos_upper[0]], [*zpos_upper, zpos_upper[0]], 
        ls='', c='w', marker='+', zorder=20, ms=4, lw=2, alpha=0.6)


# --- plotting limits --- #
delta_x = abs(ref_core[0] - ref_source[0])
delta_y = abs(ref_core[1] - ref_source[1])
delta_z = abs(ref_core[2] - ref_source[2])
xrefs, yrefs, zrefs = [ref_core[0], ref_source[0]], [ref_core[1], ref_source[1]], [ref_core[2], ref_source[2]]
factor = 1.7
xmin, xmax = min(xrefs) - delta_x * factor, max(xrefs) + delta_x * factor
ymin, ymax = min(yrefs) - delta_y * factor, max(yrefs) + delta_y * factor
zmin, zmax = min(zrefs) - delta_z * factor, max(zrefs) + delta_z * factor

# --- masks for the points --- #
core_mask = (core_x >= xmin) & (core_x <= xmax) & (core_y >= ymin) & (core_y <= ymax) & \
            (core_z >= zmin) & (core_z <= zmax)
perp_mask = (perp_x >= xmin) & (perp_x <= xmax) & (perp_y >= ymin) & (perp_y <= ymax) & \
            (perp_z >= zmin) & (perp_z <= zmax)
proj_mask = (proj_x >= xmin) & (proj_x <= xmax) & (proj_y >= ymin) & (proj_y <= ymax) & \
            (proj_z >= zmin) & (proj_z <= zmax)
source_mask = (source_x >= xmin) & (source_x <= xmax) & (source_y >= ymin) & (source_y <= ymax) & \
            (source_z >= zmin) & (source_z <= zmax)
hsource_mask = (hsource_x >= xmin) & (hsource_x <= xmax) & (hsource_y >= ymin) & (hsource_y <= ymax) & \
            (hsource_z >= zmin) & (hsource_z <= zmax)

# --- create the planes --- #
# x & y points for the planes
n_points_plane = 2
x_range = np.linspace(xmin, xmax, n_points_plane)
y_range = np.linspace(ymin, ymax, n_points_plane)
X_grid,  Y_grid  = np.meshgrid(x_range, y_range)
# reference planes
core_plane   = z_plane(X_grid, Y_grid, ref_core,   [0, 0])
perp_plane   = z_plane(X_grid, Y_grid, ref_core,   point_dir)
source_plane = z_plane(X_grid, Y_grid, ref_source, point_dir)


# --- plot the planes --- #
ax.plot_surface(X_grid, Y_grid, core_plane, alpha=0.6, color='gray', zorder=-9)
ax.plot_surface(X_grid, Y_grid, perp_plane, alpha=0.6, color='orangered', zorder=-10)
ax.plot_surface(X_grid, Y_grid, source_plane, alpha=0.6, color='royalblue', zorder=-8)

ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.zaxis.set_rotate_label(False)
ax.set_zlabel('z [m]', rotation=90)
ax.xaxis.labelpad = 10
ax.yaxis.labelpad = 10
ax.zaxis.labelpad = 10

# setting the limits
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)


# --- random sample plotting --- #
ax.plot(core_x[core_mask], core_y[core_mask], core_z[core_mask], 'o', color='dimgray', alpha=0.5, zorder=11, ms=1, label='core positions')
ax.plot(perp_x[perp_mask], perp_y[perp_mask], perp_z[perp_mask], 'o', color='r', alpha=0.5, ms=1, zorder=9, label='$\perp$-plane trajectory ext.')
ax.plot(proj_x[proj_mask], proj_y[proj_mask], proj_z[proj_mask], 'o', color='maroon', alpha=0.5, zorder=10, ms=0.5, label='$\perp$-plane projection')
ax.plot(source_x[source_mask], source_y[source_mask], source_z[source_mask], 'o', alpha=0.5, ms=1, color='b', zorder=19, label='h_max plane trajectory ext.')
ax.plot(hsource_x[hsource_mask], hsource_y[hsource_mask], hsource_z[hsource_mask], '.', color='c', zorder=0, alpha=0.5, ms=1, label='h_max positions')

ax1.hist2d(core_y, core_x, range=[np.flip(xlim), ylim], bins=150, cmap='Greys')
ax2.hist2d(source_y, source_x, range=[[1270, 1900], [-600, -100]], bins=150, cmap='Blues')
ax3.hist2d(perp_y, perp_x, range=[np.flip(xlim), ylim], bins=150, cmap='Reds')
ax4.hist2d(proj_y, proj_x, range=[np.flip(xlim), ylim], bins=150, cmap='Reds')

ax1.plot([*xpos, xpos[0]], [*ypos, ypos[0]], ls='', c='k',
         marker='+', zorder=20, ms=12, lw=2, alpha=0.6,)
ax2.plot([*ypos_upper, ypos_upper[0]], [*xpos_upper, xpos_upper[0]], ls='', c='w',
         marker='+', zorder=20, ms=12, lw=2, alpha=0.6,)
ax3.plot([*ypos_lower, ypos_lower[0]], [*xpos_lower, xpos_lower[0]], ls='', c='w',
         marker='+', zorder=20, ms=12, lw=2, alpha=0.6,)
ax4.plot([*ypos_lower, ypos_lower[0]], [*xpos_lower, xpos_lower[0]], ls='', c='w',
         marker='+', zorder=20, ms=12, lw=2, alpha=0.6,)

titles = ['core-plane', '$\perp$-upper plane', '$\perp$-lower projection', '$\perp$-lower extension']
for axe, t in zip([ax1, ax2, ax3, ax4], titles):
    axe.set_title(t)

ax3.set_xlabel('y [m]')
ax4.set_xlabel('y [m]')
ax1.set_ylabel('x [m]')
ax3.set_ylabel('y [m]')

ax.legend(loc=9)
fig.tight_layout()
ax.view_init(4, 30)
ax.set_box_aspect([1,1,1])
plt.show()
```
