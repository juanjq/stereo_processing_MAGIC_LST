```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import glob
import astropy.units as u
from astropy.coordinates import SkyCoord, AltAz
from astropy.coordinates.erfa_astrom import ErfaAstromInterpolator, erfa_astrom
from astropy.time import Time
from lstchain.reco.utils import location, add_delta_t_key, get_effective_time
from os import path

from ctapipe.coordinates import CameraFrame, TelescopeFrame

%matplotlib inline
```

```python
# NOTE: process the DL2 files first with add_inf_to_DL2.ipynb!

#dataset =  glob.glob("/fefs/aswg/workspace/abelardo.moralejo/BLLac_tuned_MC_20211028/20210808/merged/DL2_dispnorm/dl2_LST-1.Run0555*.h5")
#dataset =  glob.glob("/fefs/aswg/workspace/abelardo.moralejo/BLLac_tuned_MC_20211028/20210808/merged/DL2_dispdxdy/dl2_LST-1.Run0555*.h5")
#dataset =  glob.glob("/fefs/aswg/workspace/abelardo.moralejo/2022_MC_production/Crab_test/LZA/DL2_onlynsb_standard_RF/dl2*.h5")

dataset = glob.glob("/fefs/aswg/workspace/abelardo.moralejo/jjimenez_master/Crab_LST1/DL2/dl2_*.h5")

nominal_focal = 28 * u.m

focal = 29.30565 * u.m  # EFFECTIVE one!! (i.e. accounting for mean effect of coma aberration)
aberration_correction = focal / nominal_focal
```

```python
dataset.sort()
```

```python
tablename = "/dl2/event/telescope/parameters/LST_LSTCam"
dummy, dummy2 = [], []
t_eff, t_elapsed = 0, 0

columns=['event_id', 'reco_energy', 'reco_alt', 'reco_az', 'alt_tel', 'az_tel', 
         'reco_src_x', 'reco_src_y', 'dragon_time', 
         'concentration_pixel', 'intensity', 'ucts_trigger_type', 
         'width', 'gammaness', 'event_type', 'delta_t']

for file in dataset:
    print(file)
    tb = pd.read_hdf(file, tablename)[columns]
    lt, et = get_effective_time(tb)
    t_eff += lt
    t_elapsed += et
    dummy.append(tb)
    dummy2.append(pd.read_hdf(file, 'source_position'))
    
table = pd.concat(dummy)
table2 = pd.concat(dummy2)

# Add the source_position info:
for var in table2:
    table[var] = table2[var]


# Since we want to compare the nominal source position src_x, src_y (calculated with the nominal focal of 28 m)
# with the reconstructed direction (reco_src_x, reco_src_y), which is affected by coma aberration (light focused 
# slightly further from the center of the camera), we apply below the coma aberration to the nominal position:

table['src_x'] *= aberration_correction
table['src_y'] *= aberration_correction

dx = np.rad2deg((table['reco_src_x']-table['src_x'])/focal.to_value(u.m))
dy = np.rad2deg((table['reco_src_y']-table['src_y'])/focal.to_value(u.m))
table['theta2_on']  = (dx**2 + dy**2).astype('float32')

# Now the theta2 angle relative to the off position (opposite to the source position with respect to the 
# center of the camera):
dx = np.rad2deg((table['reco_src_x']+table['src_x'])/focal.to_value(u.m))
dy = np.rad2deg((table['reco_src_y']+table['src_y'])/focal.to_value(u.m))
table['theta2_off'] = (dx**2 + dy**2).astype('float32')
```

```python
# Select cosmics:
# v0.7: use event_type
cosmics = (table['event_type']==32)
```

```python
# Set cuts:

min_intensity = 0
max_intensity = 1e6
#max_intensity=300

ereco_min = 0 # TeV
ereco_max = 1e6 # TeV
# ereco_min = 0
# ereco_max = 0.5 # TeV


#gamma_cut=0.2   # disp_dxdy soft
#gamma_cut = 0.1844  # disp_norm soft

#gamma_cut = 0.5 # disp_dxdy
#gamma_cut = 0.4833 # disp_norm

#gamma_cut = 0.47584 # disp_dxdy  tuned to Seiya's
#gamma_cut = 0.4565 # disp_norm   tuned to seiya's

gamma_cut = 0.95 # 0.7
```

```python
nbins=100

fig = plt.figure(figsize=(16,6))


mask = ((table['gammaness']>gamma_cut) & 
        (table['intensity']>min_intensity) & 
        (table['intensity']<max_intensity) & 
        (table['reco_energy']>ereco_min) &
        (table['reco_energy']<ereco_max) &
        cosmics)

counts_on, bins  = np.histogram(table['theta2_on'][mask], bins=nbins, range=(0., 0.5))
counts_off, bins = np.histogram(table['theta2_off'][mask], bins=nbins, range=(0., 0.5))

fig.add_subplot(1, 2, 1)
plt.errorbar(0.5*(bins[1:]+bins[:-1]), counts_on, yerr=counts_on**0.5, fmt='o', ms=5)
plt.hist(table['theta2_off'][mask], bins=nbins, range=(0., 0.5), histtype='step')

plt.xlabel('$\\theta^2 (deg^2)$', fontsize=14)
plt.ylabel('Events', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0, counts_on.max()*1.15)
plt.grid()


excess = counts_on - counts_off
err = (counts_on + counts_off)**0.5


fig.add_subplot(1, 2, 2)

plt.errorbar(0.5*(bins[1:]+bins[:-1]), excess, yerr=err, fmt='o', ms=4)

plt.xlabel('$\\theta^2 (deg^2)$', fontsize=14)
plt.ylabel('Excess', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
#plt.ylim(-1500, 3500) # soft cuts
#plt.ylim(-500, 1000) # tight cuts

plt.show()


nbinscut = 6 # 18 # this defines the size of the ON region (number of bins of the theta2 histo)


non = counts_on[:nbinscut].sum()
noff = counts_off[:nbinscut].sum()

print('Excess:', non-noff, '; Off:', noff)
print('Gamma rate:', (non-noff)/t_eff.to(u.min))
print('Off rate:', noff/t_eff.to(u.min))
```

```python
from pyirf.statistics import li_ma_significance
li_ma_significance(non, noff, 1)
```

```python
time_factor = 50. / t_eff.to_value(u.h)
```

```python
noffregions = 3.
print("In 50 hours, with", noffregions, "regions:")
for crabfrac in [1, 0.1, 0.05, 0.03, 0.01]:
    print("  ", int(100*crabfrac), "% of Crab Nebula:")
    print("      S = ", li_ma_significance(time_factor*(noff+crabfrac*(non-noff)), 
                                           time_factor*noff*noffregions, 
                                           1./noffregions))
    print()
```
