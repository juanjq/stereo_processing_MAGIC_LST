```python
class StereoData:
    
    def __init__(self, df, stereo_df):
    
        # total data
        self.df = df
        
        # total data for each telescope
        self.df_tel1 = df[df.index.get_level_values('tel_id')==1]
        self.df_tel2 = df[df.index.get_level_values('tel_id')==2]
        self.df_tel3 = df[df.index.get_level_values('tel_id')==3]
        
        
        # dataframes for events in each group of telescopes 
        df12, df13, df23, df123 = df.copy(), df.copy(), df.copy(), df.copy()
        df12.query('combo_type == 1', inplace=True)
        df13.query('combo_type == 2', inplace=True)
        df23.query('combo_type == 0', inplace=True)
        df123.query('combo_type == 3', inplace=True)
        
        self.df_tels12  = df12
        self.df_tels13  = df13
        self.df_tels23  = df23
        self.df_tels123 = df123
        
        
        # dataframe for each telescope
        df1 = df123[df123.index.get_level_values('tel_id')==1]
        df2 = df123[df123.index.get_level_values('tel_id')==2]
        df3 = df123[df123.index.get_level_values('tel_id')==3]

        # extracting data
        # events 
        indexes = np.array(dl2_df.index.tolist())
        self.event_ids = indexes[1]
        self.obs_ids   = indexes[0]
        self.tel_ids   = indexes[2]
        
        indexes = np.array(df1.index.tolist())
        self.event_ids_tels = indexes[1]
        self.obs_ids_tels   = indexes[0]
        
        # energy and gammaness
        print(f'\nReading parameters...')
        self.energy     = [d['reco_energy'].to_numpy() for d in [df1, df2, df3]]
        self.gammaness = [d['gammaness'].to_numpy()   for d in [df1, df2, df3]]

        # MAGIC 1 and 2 intensities
        self.intensity = [d['intensity'].to_numpy() for d in [df1, df2, df3]]

        # X, Y direction
        self.dirX = [d['x'].to_numpy() for d in [df1, df2, df3]]
        self.dirY = [d['y'].to_numpy() for d in [df1, df2, df3]]

        # dec and right ascension
        self.dec = [d['reco_dec'].to_numpy() for d in [df1, df2, df3]]
        self.ra  = [d['reco_ra'].to_numpy()  for d in [df1, df2, df3]]
        
        # az and alt
        self.alt = [d['reco_alt'].to_numpy() for d in [df1, df2, df3]]
        self.az  = [d['reco_az'].to_numpy()  for d in [df1, df2, df3]]
        

        # energy with gammaness filtering
        print(f'\nFiltering energies...')
        self.energy_10 = [d.query('gammaness > 0.1' )['reco_energy'].to_numpy() for d in [df1, df2, df3]]
        self.energy_50 = [d.query('gammaness > 0.5' )['reco_energy'].to_numpy() for d in [df1, df2, df3]]
        self.energy_70 = [d.query('gammaness > 0.7' )['reco_energy'].to_numpy() for d in [df1, df2, df3]]
        self.energy_90 = [d.query('gammaness > 0.9' )['reco_energy'].to_numpy() for d in [df1, df2, df3]]
        self.energy_99 = [d.query('gammaness > 0.99')['reco_energy'].to_numpy() for d in [df1, df2, df3]]
        
        # energy ratios
        self.ratio_l1_m1 = self.energy[0] / self.energy[1]
        self.ratio_l1_m2 = self.energy[0] / self.energy[2]
        self.ratio_m1_m2 = self.energy[1] / self.energy[2]
        
        
        # stereo data 
        print(f'\nStereo reconstruction...\n')
        self.stereo_df = stereo_df
        
        # gammaness filters
        gfilt = [0, 0.1, 0.7, 0.8, 0.95, 0.99]
        
        # stereo parameters
        self.stereo_alt = [stereo_df.query(f'gammaness >= {g}')['reco_alt'].to_numpy() for g in gfilt]
        self.stereo_az  = [stereo_df.query(f'gammaness >= {g}')['reco_az' ].to_numpy() for g in gfilt]

        self.stereo_ra  = [stereo_df.query(f'gammaness >= {g}')['reco_ra' ].to_numpy() for g in gfilt]
        self.stereo_dec = [stereo_df.query(f'gammaness >= {g}')['reco_dec'].to_numpy() for g in gfilt]

        self.stereo_p_alt = [stereo_df.query(f'gammaness >= {g}')['pointing_alt'].to_numpy() for g in gfilt]
        self.stereo_p_az  = [stereo_df.query(f'gammaness >= {g}')['pointing_az' ].to_numpy() for g in gfilt]

        self.stereo_p_ra  = [stereo_df.query(f'gammaness >= {g}')['pointing_ra' ].to_numpy() for g in gfilt]
        self.stereo_p_dec = [stereo_df.query(f'gammaness >= {g}')['pointing_dec'].to_numpy() for g in gfilt]

        self.stereo_energy    = [stereo_df.query(f'gammaness >={g}')['reco_energy'].to_numpy() for g in gfilt]
        self.stereo_gammaness = [stereo_df.query(f'gammaness >={g}')['gammaness'  ].to_numpy() for g in gfilt]
```
