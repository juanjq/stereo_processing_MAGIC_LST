## Import packages

```python
# --- packages --- #
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import sys, os
import matplotlib.colors as colors

# --- scripts paths --- #
sys.path.insert(0, '/fefs/aswg/workspace/juan.jimenez/stereo_analysis/scripts')
import auxiliar as aux
# graphic parameters
aux.params()

# --- file paths --- #
dir_stereo_reco   = '/fefs/aswg/workspace/juan.jimenez/data/datasets/'
# filenames
name_stereo_3T    = 'stereo_data_3T'
name_stereo_total = 'stereo_data_total'
name_dl2_lst_mcp  = 'stereo_data_LST'
name_stereo_magic = 'stereo_data_MAGIC'
name_dl2_lst_lst  = 'dl2_data_LST'
name_dl2_magic    = 'dl2_data_MAGIC'
```

## Reading all dataframes

```python
# reading all data already produced
df_total      = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_total}.h5'))
df_t1 = df_total.query(f'tel_id == 1')
df_t2 = df_total.query(f'tel_id == 2')
df_t3 = df_total.query(f'tel_id == 3')

df_3T_mcp     = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_3T}.h5'))
df_lst_mcp    = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_lst_mcp}.h5'))
df_magic_mcp  = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_magic}.h5'))
df_lst_lst    = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_lst_lst}.h5'))
df_magic_mars = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_magic}.h5'))
df_magic_mars.query(f'reco_energy != -0.001', inplace=True)
```

```python
ids_mars = df_magic_mars['total_id'].to_numpy()

df_3T_mcp_red     = df_3T_mcp.query(   f'total_id in @ids_mars', inplace=False)
df_lst_mcp_red    = df_lst_mcp.query(  f'total_id in @ids_mars', inplace=False)
df_magic_mcp_red  = df_magic_mcp.query(f'total_id in @ids_mars', inplace=False)
df_lst_lst_red    = df_lst_lst.query(  f'total_id in @ids_mars', inplace=False)
```

## Extracting the values from the dataframes

```python
# energy cuts
ECuts = [1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3]

# gammaness of each dataframe filtered in energy
g_3T_mcp     = df_3T_mcp['gammaness'].to_numpy()
G_3T_mcp     = [df_3T_mcp.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['gammaness'].to_numpy() for i in range(len(ECuts)-1)]
g_magic_mcp  = df_magic_mcp['gammaness'].to_numpy()
G_magic_mcp  = [df_magic_mcp.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['gammaness'].to_numpy() for i in range(len(ECuts)-1)]
g_magic_mars = df_magic_mars['gammaness'].to_numpy()
G_magic_mars = [df_magic_mars.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['gammaness'].to_numpy() for i in range(len(ECuts)-1)]
g_lst_mcp    = df_lst_mcp['gammaness'].to_numpy()
G_lst_mcp    = [df_lst_mcp.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['gammaness'].to_numpy() for i in range(len(ECuts)-1)]
g_lst_lst    = df_lst_lst['gammaness'].to_numpy()
G_lst_lst    = [df_lst_lst.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['gammaness'].to_numpy() for i in range(len(ECuts)-1)]

# intensity of each dataframe filtered in energy
i_magic_mars = df_magic_mars['intensity_M1'].to_numpy()
I_magic_mars = [df_magic_mars.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['intensity_M1'].to_numpy() for i in range(len(ECuts)-1)]
i_lst_lst    = df_lst_lst['intensity'].to_numpy()
I_lst_lst    = [df_lst_lst.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['intensity'].to_numpy() for i in range(len(ECuts)-1)]

i_LST = df_t1['intensity'].to_numpy()
I_LST = [df_t1.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['intensity'].to_numpy() for i in range(len(ECuts)-1)]
i_M1  = df_t2['intensity'].to_numpy()
I_M1  = [df_t2.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['intensity'].to_numpy() for i in range(len(ECuts)-1)]
i_M2  = df_t3['intensity'].to_numpy()
I_M2  = [df_t3.query(f'reco_energy > {ECuts[i]} & reco_energy < {ECuts[i+1]}')['intensity'].to_numpy() for i in range(len(ECuts)-1)]


# computing ratios
r_LST_lst_mcp       = g_lst_lst / g_lst_mcp
r_3T_LST_mcp_lst    = g_3T_mcp  / g_lst_lst
r_3T_MAGIC_mcp_mcp  = g_3T_mcp  / g_magic_mcp
r_LST_MAGIC_mcp_mcp = g_lst_mcp / g_magic_mcp
# dropping nan
r_LST_lst_mcp       = r_LST_lst_mcp[~np.isnan(r_LST_lst_mcp)]
r_3T_LST_mcp_lst    = r_3T_LST_mcp_lst[~np.isnan(r_3T_LST_mcp_lst)]
r_3T_MAGIC_mcp_mcp  = r_3T_MAGIC_mcp_mcp[~np.isnan(r_3T_MAGIC_mcp_mcp)]
r_LST_MAGIC_mcp_mcp = r_LST_MAGIC_mcp_mcp[~np.isnan(r_LST_MAGIC_mcp_mcp)]
# mars with filtering
```

## Gammaness distributions for all data types

```python
# binning
bins = np.linspace(0, 1, 120)

fig, ax = plt.subplots(figsize=(10,6))
ax.set_title(f'Energy reconstruction {len(g_3T_mcp)} events')

ax.hist(g_3T_mcp, bins, color='c', alpha=1, histtype='stepfilled', label='stereo-3T')
ax.hist(g_lst_lst, bins, histtype='step',alpha=1, lw=3, color='deeppink', label='LST-lstchain')
ax.hist(g_lst_mcp, bins, histtype='step', lw=3, color='gold', label='LST-mcp')
ax.hist(g_magic_mcp, bins, alpha=1, color='darkblue', histtype='step', lw=3, label='MAGIC-mcp')
ax.hist(g_magic_mars, bins, alpha=1, color='crimson', histtype='step', lw=3, label='MAGIC-MARS')

ax.set_xlabel('gammaness')
ax.set_ylabel('counts')
ax.legend(loc=9)
ax.set_yscale('log')
plt.show()
```

## Gammaness comparison between datasets

```python
# --- energy comparison --- #
bins = np.linspace(0, 1, 80)
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(9.5,8.8), sharex=True, sharey=True)
fig.suptitle(f'Gammaness reconstruction')

ax1.hist2d(g_3T_mcp, g_magic_mcp,bins=bins, cmap='inferno', norm=colors.LogNorm())
ax2.hist2d(g_lst_lst, g_lst_mcp, bins=bins, cmap='inferno', norm=colors.LogNorm())
ax3.hist2d(g_3T_mcp, g_lst_lst, bins=bins, cmap='inferno', norm=colors.LogNorm())
ax4.hist2d(g_lst_mcp,g_magic_mcp, bins=bins, cmap='inferno', norm=colors.LogNorm())

ax1.set_title(f'3Tel mcp & MAGIC mcp')
ax2.set_title(f'LST; lstchain & mcp')
ax3.set_title(f'3Tel mcp & LST lstchain')
ax4.set_title(f'LST mcp & MAGIC mcp')

ax1.set_xlabel('$\\gamma\ $(M-mars)')
ax1.set_ylabel('$\\gamma\ $(M-mcp)')
ax2.set_xlabel('$\\gamma\ $(LST-lst)')
ax2.set_ylabel('$\\gamma\ $(LST-mcp)')
ax3.set_xlabel('$\\gamma\ $(3T-mcp)')
ax3.set_ylabel('$\\gamma\ $(LST-lst)')
ax4.set_xlabel('$\\gamma\ $(LST-mcp)')
ax4.set_ylabel('$\\gamma\ $(M-mcp)')

for ax in [ax1, ax2, ax3, ax4]:
    ax.plot([min(bins),max(bins)], [min(bins),max(bins)], lw=1 ,ls='--', c='w')
    ax.set_facecolor('k')
fig.tight_layout()
plt.show()

# --- energy ratios --- #
bins = np.logspace(-5.5, 5.5, 120)

fig, ax = plt.subplots(figsize=(10.5,5))
ax.set_title(f'Gammaness ratios')

ax.hist(r_3T_LST_mcp_lst, bins=bins, label='3T(mcp)/LST(lst)', histtype='stepfilled', color='c')
ax.hist(r_3T_MAGIC_mcp_mcp, bins=bins, label='3T(mcp)/M(mcp)', histtype='step', lw=2, color='deeppink')
ax.hist(r_LST_lst_mcp, bins=bins, label='LST(mcp)/LST(lst)', histtype='step', lw=2, color='darkblue')
ax.hist(r_LST_MAGIC_mcp_mcp, bins=bins, label='LST(mcp)/MAGIC(mcp)', histtype='step', lw=2, color='gold')

ax.axvline(1, color='k', lw=1, linestyle='--')

ax.legend(loc=2)
ax.loglog()
ax.set_xlabel('$\\gamma_i/\\gamma_j$')
ax.set_ylabel('counts')
plt.show()
```
