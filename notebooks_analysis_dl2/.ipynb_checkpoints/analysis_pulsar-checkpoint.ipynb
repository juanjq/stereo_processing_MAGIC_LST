```python
df_T1 = df_merged.query('tel_id == 1', inplace=False)
df_T2 = df_merged.query('tel_id == 2', inplace=False)
df_T3 = df_merged.query('tel_id == 3', inplace=False)

dfmc_T1 = df_mc_tot.query('tel_id == 1', inplace=False)
dfmc_T2 = df_mc_tot.query('tel_id == 2', inplace=False)
dfmc_T3 = df_mc_tot.query('tel_id == 3', inplace=False)
```

```python
df_T1['time_diff'] = df_T1['timestamp'].diff()

df = df_T1.dropna()

df = df[df['time_diff'] <= 1]

fft_vals = np.fft.fft(df['time_diff'])

power_spectrum = np.abs(fft_vals) ** 2

sampling_rate = 1 / np.mean(df['time_diff'])
frequencies = np.fft.fftfreq(len(df['time_diff']), d=1/sampling_rate)
```

```python

frequencies[power_spectrum>8000] = np.nan


# Step 8: Plot the power spectrum
plt.plot(frequencies, power_spectrum, lw=2)
plt.xlabel('Frequency')
plt.ylabel('Power Spectrum')
plt.ylim(-1e2, 8e3)
plt.show()
```

```python

frequencies[frequencies>8000] = np.nan

# Step 9: Identify the dominant frequency or peak in the power spectrum
dominant_frequency = frequencies[np.argmax(power_spectrum)]
estimated_period = 1 / dominant_frequency
print(f"Estimated period: {estimated_period}")
```
