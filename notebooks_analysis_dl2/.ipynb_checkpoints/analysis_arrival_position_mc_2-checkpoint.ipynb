## Packages and definition of parameters

```python
# --- packages --- #
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import os, sys, glob
import matplotlib.colors as colors

from ctapipe.instrument import SubarrayDescription

# --- path to all scripts --- #
sys.path.insert(0, '/fefs/aswg/workspace/juan.jimenez/stereo_analysis/scripts')
import auxiliar as aux
import geometry as geom
# graphic parameters
aux.params()

# -- image fitting limits [m] for the maps --- #
ylim  = (-110, 110)
xlim  = (-90,   90)

# --- paths --- #
dir_mask = '/fefs/aswg/LST1MAGIC/mc/DL2/ST0316A/NSB1.5/GammaTest/v01.2/dec_2276/*.h5'
event_coinc_dir   = '/fefs/aswg/workspace/juan.jimenez/data/dl1/joint_runs_magic_lst'
dir_stereo_reco   = '/fefs/aswg/workspace/juan.jimenez/data/datasets/'
name_stereo_3T    = 'stereo_data_3T'
name_stereo_total = 'stereo_data_total'

# text for labeling telescopes
tel_lab = ['LST-1', 'MAGIC-1', 'MAGIC-2'] 

# dataframe masking function
def mask_fun(x, y, z, lims):
    
    xmin, xmax, ymin, ymax, zmin, zmax = lims
    
    mask = (x >= xmin) & (x <= xmax) & (y >= ymin) & (y <= ymax) & (z >= zmin) & (z <= zmax)
    return mask
# function to generate a range for the histogram
def rangehist(x, y, sigma=1):
    mhx, mhy = np.mean(x), np.mean(y)
    sx,  sy  = np.std(x), np.std(y)

    return [[mhx - sx * sigma, mhx + sx * sigma], [mhy - sy * sigma, mhy + sy * sigma]]
# finding limits for the data
def lim_plot(ref_lower, ref_upper, factor):
    delta_x, delta_y, delta_z = [abs(ref_lower[i] - ref_upper[i]) for i in range(len(ref_lower))]
    
    xrefs = [ref_lower[0], ref_upper[0]]
    yrefs = [ref_lower[1], ref_upper[1]]
    zrefs = [ref_lower[1], ref_upper[1]]
    
    xmin, xmax = min(xrefs) - delta_x * factor, max(xrefs) + delta_x * factor
    ymin, ymax = min(yrefs) - delta_y * factor, max(yrefs) + delta_y * factor
    zmin, zmax = min(zrefs) - delta_z * factor, max(zrefs) + delta_z * factor

    return xmin, xmax, ymin, ymax, zmin, zmax
```

## Random projection test

```python
# --- pointing direction --- #
point_zd, point_az = 25, 45
point_dir = (point_zd, point_az)
print(f'The pointing direction is zd = {point_zd:.2f}fig.tight_layout()deg and az = {point_az:.2f}deg')

# --- defining the random variables --- #
# number of points and altitude mean
N, mean_h = 300, 10
zd = np.random.normal(point_zd, 7, N)
az = np.random.normal(point_az, 7, N)

h_max  = np.random.normal(mean_h, 3, N)

core_x = np.random.normal(0, 5, N)
core_y = np.random.normal(0, 5, N)
core_z = np.zeros(N)


# --- reference points --- #
ref_lower = np.array([0, 0, 0])
ref_upper = geom.z_projection(*ref_lower, *point_dir, np.mean(h_max))
print(f'\nRef point for cores is  [{ref_lower[0]:.2f}, {ref_lower[1]:.2f}, {ref_lower[2]:.2f}]m')
print(f'Ref point for source is [{ref_upper[0]:.2f}, {ref_upper[1]:.2f}, {ref_upper[2]:.2f}]m')


# --- defining the dataframe --- #
MATRIX = {'reco_zd':zd, 'reco_az':az, 'h_max':h_max, 'core_x':core_x, 'core_y':core_y, 'core_z':core_z}
data = pd.DataFrame(MATRIX)

#########################################
# --- calculating plane projections --- #
#########################################
cores_df      = [data['core_x'], data['core_y'], data['core_z']]
directions_df = [data['reco_zd'], data['reco_az']]

lower_points  = geom.plane_projection(ref_lower, point_dir, *cores_df, *directions_df)
proj_points   = geom.plane_projection(ref_lower, point_dir, *cores_df, *point_dir)
upper_points  = geom.plane_projection(ref_upper, point_dir, *cores_df, *directions_df)
shower_points = geom.z_projection(*cores_df, *directions_df, data['h_max'])

data['lower_x'],  data['lower_y'],  data['lower_z']  = lower_points
data['proj_x'],   data['proj_y'],   data['proj_z']   = proj_points
data['upper_x'],  data['upper_y'],  data['upper_z']  = upper_points
data['shower_x'], data['shower_y'], data['shower_z'] = shower_points

lower_x, lower_y, lower_z    = lower_points
proj_x, proj_y, proj_z       = proj_points
upper_x, upper_y, upper_z    = upper_points
shower_x, shower_y, shower_z = shower_points
#########################################

# --- plotting limits --- #
lims = lim_plot(ref_lower, ref_upper, factor=2)
xmin, xmax, ymin, ymax, zmin, zmax = lims
print(f'\nThe plotting limits that will be used are:')
print(f'xlims:[{xmin:.1f},{xmax:.1f}], ylims:[{ymin:.1f},{ymax:.1f}], zlims:[{zmin:.1f},{zmax:.1f}]')

# --- masks for the points --- #
xmin, xmax, ymin, ymax, zmin, zmax
core_mask   = mask_fun(core_x,   core_y,   core_z,  lims)
proj_mask   = mask_fun(proj_x,   proj_y,   proj_z,   lims)
upper_mask  = mask_fun(upper_x,  upper_y,  upper_z,  lims)
lower_mask  = mask_fun(lower_x,  lower_y,  lower_z,  lims)
shower_mask = mask_fun(shower_x, shower_y, shower_z, lims)

display(data.head(7))
```

```python
fig = plt.figure(figsize=(10,10))
ax  = fig.add_subplot(1, 1, 1, projection='3d') 

# --- plot the reference points --- #
ref_points = [[ref_lower[i], ref_upper[i]] for i in range(len(ref_lower))]
ax.plot(*ref_points, '--', c='k', marker='o', ms=6, lw=2, zorder=15)

# --- create the planes --- #
# x & y points for the planes
n_points_plane = 2
x_range = np.linspace(xmin, xmax, n_points_plane)
y_range = np.linspace(ymin, ymax, n_points_plane)
X_grid,  Y_grid  = np.meshgrid(x_range, y_range)
# reference planes
core_plane  = geom.z_plane(X_grid, Y_grid, ref_lower, [0, 0])
lower_plane = geom.z_plane(X_grid, Y_grid, ref_lower, point_dir)
upper_plane = geom.z_plane(X_grid, Y_grid, ref_upper, point_dir)

# --- plot the planes --- #
ax.plot_surface(X_grid, Y_grid, core_plane,   alpha=0.4, color='gray', zorder=-9)
ax.plot_surface(X_grid, Y_grid, lower_plane,   alpha=0.4, color='r', zorder=-10)
ax.plot_surface(X_grid, Y_grid, upper_plane, alpha=0.4, color='b', zorder=-8)

# --- random sample plotting --- #
cores_masked  = core_x[core_mask], core_y[core_mask], core_z[core_mask]
proj_masked   = proj_x[proj_mask], proj_y[proj_mask], proj_z[proj_mask]
lower_masked  = lower_x[lower_mask], lower_y[lower_mask], lower_z[lower_mask]
upper_masked  = upper_x[upper_mask], upper_y[upper_mask], upper_z[upper_mask]
shower_masked = shower_x[shower_mask], shower_y[shower_mask], shower_z[shower_mask]

ax.plot(*cores_masked, 'o', color='gray', zorder=11, label='core positions')
ax.plot(*proj_masked, '.', color='maroon', zorder=10, ms=2, label='$\perp$-plane projection')
ax.plot(*lower_masked, 'o', color='r', zorder=9, label='$\perp$-plane trajectory ext.')
ax.plot(*upper_masked, 'o', color='b', zorder=19, label='h_max plane trajectory ext.')
ax.plot(*shower_masked, '.', color='c', zorder=19, label='h_max positions')

# setting the limits
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

# other parameters
ax.legend(loc=9)
ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.zaxis.set_rotate_label(False)
ax.set_zlabel('z [m]', rotation=90)
fig.tight_layout()
ax.set_box_aspect([1,1,1])

ax.view_init(4, 30)
plt.show()
```

## Reading real data

```python
#### --- extracting pointing direction --- #
point_zd  = 90 - np.rad2deg(df['pointing_alt'].to_numpy()[0])
point_az  = np.rad2deg(df['pointing_az'].to_numpy()[0])
point_dir = (point_zd, point_az)
print(f'The pointing direction is zd = {point_zd:.2f}deg and az = {point_az:.2f}deg')

############################################
# --- reading the given main variables --- #
############################################
print(f'\nReading the variables form the dataframe...')
# reconstructed directions
df['reco_zd'] = 90 - df['reco_alt']
zd, az = df['reco_zd'].to_numpy(), df['reco_az'].to_numpy()
# true directions
df['true_zd'] = 90 - df['true_alt']
true_zd, true_az = df['true_zd'].to_numpy(), df['true_az'].to_numpy()

# reconstructed core positions
core_x = df['core_y'].to_numpy() * -1
core_y = df['core_x'].to_numpy()
core_z = np.zeros(len(core_x))
df['core_z'] = core_z
# true cores positions
true_core_x = df['true_core_y'].to_numpy() * -1
true_core_y = df['true_core_x'].to_numpy()
true_core_z = core_z
df['true_core_z'] = true_core_z

# altitude of maximum intensity
h_max = df['h_max'].to_numpy()
############################################

# --- reference points --- #
ref_lower = np.array([0, 0, 0])
ref_upper = geom.z_projection(*ref_lower, *point_dir, np.mean(h_max))
print(f'\nRef point for lower plane is [{ref_lower[0]:.1f}, {ref_lower[1]:.1f}, {ref_lower[2]:.1f}]m')
print(f'Ref point for upper plane is [{ref_upper[0]:.1f}, {ref_upper[1]:.1f}, {ref_upper[2]:.1f}]m')

# --- telescopes projections --- #
print(f'\nCalculating the telescope projection positions...')
t1_lower = geom.plane_projection(ref_lower, point_dir, xpos[0], ypos[0], zpos[0], point_zd, point_az)
t1_upper = geom.plane_projection(ref_upper, point_dir, xpos[0], ypos[0], zpos[0], point_zd, point_az)
t2_lower = geom.plane_projection(ref_lower, point_dir, xpos[1], ypos[1], zpos[1], point_zd, point_az)
t2_upper = geom.plane_projection(ref_upper, point_dir, xpos[1], ypos[1], zpos[1], point_zd, point_az)
t3_lower = geom.plane_projection(ref_lower, point_dir, xpos[2], ypos[2], zpos[2], point_zd, point_az)
t3_upper = geom.plane_projection(ref_upper, point_dir, xpos[2], ypos[2], zpos[2], point_zd, point_az)
xpos_lower, xpos_upper = [t1_lower[0], t2_lower[0], t3_lower[0]], [t1_upper[0], t2_upper[0], t3_upper[0]]
ypos_lower, ypos_upper = [t1_lower[1], t2_lower[1], t3_lower[1]], [t1_upper[1], t2_upper[1], t3_upper[1]]
zpos_lower, zpos_upper = [t1_lower[2], t2_lower[2], t3_lower[2]], [t1_upper[2], t2_upper[2], t3_upper[2]]


#########################################
# --- calculating plane projections --- #
#########################################
print('\nProjecting the points in the planes...')
cores_df      = [core_x, core_y, core_z]
true_cores_df = [true_core_x, true_core_y, true_core_z]
dirs_df       = [zd, az]
true_dirs_df  = [true_zd, true_az]

lower_points  = geom.plane_projection(ref_lower, point_dir, *cores_df, *dirs_df)
proj_points   = geom.plane_projection(ref_lower, point_dir, *cores_df, *point_dir)
upper_points  = geom.plane_projection(ref_upper, point_dir, *cores_df, *dirs_df)
shower_points = geom.z_projection(*cores_df, *dirs_df, h_max)
true_lower_points  = geom.plane_projection(ref_lower, point_dir, *true_cores_df, *true_dirs_df)
true_proj_points   = geom.plane_projection(ref_lower, point_dir, *true_cores_df, *point_dir)
true_upper_points  = geom.plane_projection(ref_upper, point_dir, *true_cores_df, *true_dirs_df)

df['lower_x'],  df['lower_y'],  df['lower_z']  = lower_points
df['proj_x'],   df['proj_y'],   df['proj_z']   = proj_points
df['upper_x'],  df['upper_y'],  df['upper_z']  = upper_points
df['shower_x'], df['shower_y'], df['shower_z'] = shower_points
df['true_lower_x'],  df['true_lower_y'],  df['true_lower_z']  = true_lower_points
df['true_proj_x'],   df['true_proj_y'],   df['true_proj_z']   = true_proj_points
df['true_upper_x'],  df['true_upper_y'],  df['true_upper_z']  = true_upper_points

lower_x,  lower_y,  lower_z  = lower_points
proj_x,   proj_y,   proj_z   = proj_points
upper_x,  upper_y,  upper_z  = upper_points
shower_x, shower_y, shower_z = shower_points
true_lower_x, true_lower_y, true_lower_z = true_lower_points
true_proj_x,  true_proj_y,  true_proj_z  = true_proj_points
true_upper_x, true_upper_y, true_upper_z = true_upper_points
#########################################


#########################################
# ------- calculating distances ------- #
#########################################
cores_df_real      = [core_y,      core_x * -1,      core_z]
true_cores_df_real = [true_core_y, true_core_x * -1, true_core_z]
lower_points_real  = geom.plane_projection(ref_lower, point_dir, *cores_df_real, *dirs_df)
proj_points_real   = geom.plane_projection(ref_lower, point_dir, *cores_df_real, *point_dir)
upper_points_real  = geom.plane_projection(ref_upper, point_dir, *cores_df_real, *dirs_df)
shower_points_real = geom.z_projection(*cores_df, *dirs_df, h_max)
true_lower_points_real  = geom.plane_projection(ref_lower, point_dir, *true_cores_df_real, *true_dirs_df)
true_proj_points_real   = geom.plane_projection(ref_lower, point_dir, *true_cores_df_real, *point_dir)
true_upper_points_real  = geom.plane_projection(ref_upper, point_dir, *true_cores_df_real, *true_dirs_df)

print('\nCalculating distances and impact parameters...')
df['dist_core']  = geom.dist3d(*cores_df_real,     *true_cores_df_real)
df['dist_proj']  = geom.dist3d(*proj_points_real,  *true_proj_points_real)
df['dist_lower'] = geom.dist3d(*lower_points_real, *true_lower_points_real)
df['dist_upper'] = geom.dist3d(*upper_points_real, *true_upper_points_real)
#########################################

# --- recalculating impact parameter --- #
df['recal_impact_T1']      = geom.dist3d(*cores_df_real, *coord_t1)
df['recal_impact_T2']      = geom.dist3d(*cores_df_real, *coord_t2)
df['recal_impact_T3']      = geom.dist3d(*cores_df_real, *coord_t3)
df['recal_true_impact_T1'] = geom.dist3d(*true_cores_df_real, *coord_t1)
df['recal_true_impact_T2'] = geom.dist3d(*true_cores_df_real, *coord_t2)
df['recal_true_impact_T3'] = geom.dist3d(*true_cores_df_real, *coord_t3)
df['error_impact'] = np.abs(df['recal_impact_T1']-df['impact'])


# --- plotting limits --- #
lims_reco = lim_plot(ref_lower, ref_upper, factor=2)
lims_true = lim_plot(ref_lower, ref_upper, factor=1)
xminR, xmaxR, yminR, ymaxR, zminR, zmaxR = lims_reco
xminT, xmaxT, yminT, ymaxT, zminT, zmaxT = lims_true
print(f'\nThe plotting limits that will be used in the reconstruced data:')
print(f'xlims:[{xminR:.1f},{xmaxR:.1f}], ylims:[{yminR:.1f},{ymaxR:.1f}], zlims:[{zminR:.1f},{zmaxR:.1f}]')
print(f'\nThe plotting limits that will be used in the true data:')
print(f'xlims:[{xminT:.1f},{xmaxT:.1f}], ylims:[{yminT:.1f},{ymaxT:.1f}], zlims:[{zminT:.1f},{zmaxT:.1f}]')


# --- masks for the points --- #
print(f'\nMasking the points within the limits...')
core_mask   = mask_fun(core_x,   core_y,   core_z,   lims_reco)
proj_mask   = mask_fun(proj_x,   proj_y,   proj_z,   lims_reco)
upper_mask  = mask_fun(upper_x,  upper_y,  upper_z,  lims_reco)
lower_mask  = mask_fun(lower_x,  lower_y,  lower_z,  lims_reco)
shower_mask = mask_fun(shower_x, shower_y, shower_z, lims_reco)
true_core_mask  = mask_fun(true_core_x,  true_core_y,  true_core_z,  lims_true)
true_proj_mask  = mask_fun(true_proj_x,  true_proj_y,  true_proj_z,  lims_true)
true_upper_mask = mask_fun(true_upper_x, true_upper_y, true_upper_z, lims_true)
true_lower_mask = mask_fun(true_lower_x, true_lower_y, true_lower_z, lims_true)

display(df.head(7))
```

## Checking impact parameter recalculation

$$impact = \left |\begin{pmatrix}core_x \\ core_y\\ core_z=0\end{pmatrix}-\begin{pmatrix}tel_x \\ tel_y\\ tel_z\end{pmatrix}\right |$$

$$impact = \left |\begin{pmatrix}core_x \\ core_y\\ core_z=0\end{pmatrix}-\begin{pmatrix}tel_x.core \\ tel_y.core\\ tel_z.core =0\end{pmatrix}\right |$$

### - Impact ratios

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

range_1 = (0.75, 1.1)
range_2 = (0.25, 1.75)

ratio_t1 = df['impact'] / df['recal_impact_T1']
ax1.hist(ratio_t1, bins=150, range=range_1, histtype='stepfilled', color='c', label='impact LST-1')
ratio_t2 = dfT2['impact'].to_numpy() / df['recal_impact_T2'].to_numpy()
ax1.hist(ratio_t2, bins=150, range=range_1, histtype='step', lw=3, color='darkblue', label='impact M1')
ratio_t3 = dfT3['impact'].to_numpy() / df['recal_impact_T3'].to_numpy()
ax1.hist(ratio_t3, bins=150, range=range_1, histtype='step', lw=3, color='deeppink', label='impact M2')

ratio_t1 = df['true_impact'] / df['recal_true_impact_T1']
ax2.hist(ratio_t1, bins=150, range=range_2, histtype='stepfilled', color='c', label='impact LST-1')
ratio_t2 = dfT2['true_impact'].to_numpy() / df['recal_true_impact_T2'].to_numpy()
ax2.hist(ratio_t2, bins=150, range=range_2, histtype='step', lw=3, color='darkblue', label='impact M1')
ratio_t3 = dfT3['true_impact'].to_numpy() / df['recal_true_impact_T3'].to_numpy()
ax2.hist(ratio_t3, bins=150, range=range_2, histtype='step', lw=3, color='deeppink', label='impact M2')

ax1.set_xlim(*range_1)
ax2.set_xlim(*range_2)

ax1.set_title(f'Impact parameter ratios')
ax2.set_title(f'True impact parameter ratios')
for ax, loc in zip([ax1, ax2], [2,1]):
    ax.set_xlabel('impact/recalc_impact')
    ax.legend(loc=loc)
ax1.set_ylabel('counts')
fig.tight_layout()
plt.show()
```

## Distance and positions

```python
# --- binning the data in positions --- #
ylim  = (-110, 110)
xlim  = (-90,   90)

Nbins = 100
bins_x = np.linspace(*xlim, Nbins)
bins_y = np.linspace(*ylim, Nbins)

bins_x_center = (bins_x[:-1] + bins_x[1:]) / 2
bins_y_center = (bins_y[:-1] + bins_y[1:]) / 2

df['bin_x'] = pd.cut(df['core_x'], bins=bins_y, include_lowest=True)
df['bin_y'] = pd.cut(df['core_y'], bins=bins_x, include_lowest=True)

binned_dist_core    = df.groupby(['bin_x', 'bin_y'])['dist_core'].mean().unstack()
binned_dist_lower   = df.groupby(['bin_x', 'bin_y'])['dist_lower'].mean().unstack()
binned_dist_upper   = df.groupby(['bin_x', 'bin_y'])['dist_upper'].mean().unstack()
binned_error_impact = df.groupby(['bin_x', 'bin_y'])['error_impact'].mean().unstack()
binned_h_max        = df.groupby(['bin_x', 'bin_y'])['h_max'].mean().unstack()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 9), sharex=True, sharey=True)

# plot the binned data
cmesh1 = ax1.pcolormesh(bins_x_center, bins_y_center, binned_dist_core,    cmap='jet')
fig.colorbar(cmesh1, ax=ax1, label='distance true-reco [m]')
cmesh2 = ax2.pcolormesh(bins_x_center, bins_y_center, binned_dist_upper,   cmap='jet')
fig.colorbar(cmesh2, ax=ax2, label='distance true-reco [m]')
cmesh3 = ax3.pcolormesh(bins_x_center, bins_y_center, binned_error_impact, cmap='jet')
fig.colorbar(cmesh2, ax=ax3, label='|impact-recalc_impact| [m]')
cmesh4 = ax4.pcolormesh(bins_x_center, bins_y_center, binned_h_max,        cmap='jet')
fig.colorbar(cmesh4, ax=ax4, label='h_max [m]')

# plotting directions
init_lower = ref_lower[:-1]
init_upper = ref_upper[:-1]
ax1.plot(*geom.line_in_direction(point_az, init_lower, ax1), ls='--', lw=2, c='k')
ax2.plot(*geom.line_in_direction(point_az, init_lower, ax2), ls='--', lw=2, c='k')
ax3.plot(*geom.line_in_direction(point_az, init_lower, ax3), ls='--', lw=2, c='k')
ax4.plot(*geom.line_in_direction(point_az, init_lower, ax4), ls='--', lw=2, c='k')

titles = ['true-reco dist (core-plane)', 'true-reco dist ($\perp$-plane)', 
          'error in impact param recalc', 'h_max']

for ax, t in zip([ax1, ax2, ax3, ax4], titles):
    ax.set_title(t)
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.plot(xpos, ypos, ls='', c='w', marker='+', ms=15, lw=20)

ax3.set_xlabel('core_x [m]')
ax4.set_xlabel('core_x [m]')
ax1.set_ylabel('core_y [m]')
ax3.set_ylabel('core_y [m]')

fig.tight_layout()
plt.show()
```
