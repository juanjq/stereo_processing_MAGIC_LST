## Import packages

```python
# --- packages --- #
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import sys, os
pd.set_option("display.max_columns", None)

from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))


# --- scripts paths --- #
sys.path.insert(0, '/fefs/aswg/workspace/juan.jimenez/stereo_analysis/scripts')
import auxiliar as aux
# graphic parameters
aux.params()

# --- file paths --- #
dir_stereo_reco   = '/fefs/aswg/workspace/juan.jimenez/data/datasets/'
# filenames
name_stereo_3T    = 'stereo_data_3T'
name_stereo_total = 'stereo_data_total'
name_dl2_lst_mcp  = 'stereo_data_LST'
name_stereo_magic = 'stereo_data_MAGIC'
name_dl2_lst_lst  = 'dl2_data_LST'
name_dl2_magic    = 'dl2_data_MAGIC'
```

## Reading all dataframes

```python
# reading all data already produced
df_total      = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_total}.h5'))
df_3T_mcp     = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_3T}.h5'))
df_lst_mcp    = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_lst_mcp}.h5'))
df_magic_mcp  = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_magic}.h5'))
df_lst_lst    = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_lst_lst}.h5'))
df_magic_mars = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_magic}.h5'))
df_magic_mars.query(f'reco_energy != -0.001', inplace=True)

df_3T_intensity = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_3T}_intensity.h5'))
df_3T_simple    = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_3T}_simple.h5'))
```

```python
ids_mars = df_magic_mars['total_id'].to_numpy()

df_3T_mcp_red     = df_3T_mcp.query(   f'total_id in @ids_mars', inplace=False)
df_lst_mcp_red    = df_lst_mcp.query(  f'total_id in @ids_mars', inplace=False)
df_magic_mcp_red  = df_magic_mcp.query(f'total_id in @ids_mars', inplace=False)
df_lst_lst_red    = df_lst_lst.query(  f'total_id in @ids_mars', inplace=False)
```

## Extracting the values from the dataframes

```python
# gammanes values for the filters
gammas = [0., 0.1, 0.5, 0.7, 0.8, 0.9, 0.99]

# energy of each dataframe filtered in gammaness
E_3T_mcp     = [df_3T_mcp.query(    f'gammaness > {g}')['reco_energy'].to_numpy() for g in gammas]
E_magic_mcp  = [df_magic_mcp.query( f'gammaness > {g}')['reco_energy'].to_numpy() for g in gammas]
E_magic_mars = [df_magic_mars.query(f'gammaness > {g}')['reco_energy'].to_numpy() for g in gammas]
E_lst_mcp    = [df_lst_mcp.query(   f'gammaness > {g}')['reco_energy'].to_numpy() for g in gammas]
E_lst_lst    = [df_lst_lst.query(   f'gammaness > {g}')['reco_energy'].to_numpy() for g in gammas]

E_3T_intensity = [df_3T_intensity.query(f'gammaness > {g}')['reco_energy'].to_numpy() for g in gammas]
E_3T_simple    = [df_3T_mcp.query(      f'gammaness > {g}')['reco_energy'].to_numpy() for g in gammas]

# now we can filter in gamma but with the threshold given by 3-T stereo analysis with magic-mcp-pipe
dfs_3T  = [df_3T_mcp.query(f'gammaness > {g}', inplace=False) for g in gammas]
eventsF = [df['total_id'].to_numpy() for df in dfs_3T]

# filtering same events that the 3Tel stereo filtered in gammaness
E_3T_mcpF     = E_3T_mcp.copy()
E_magic_mcpF  = [df_magic_mcp.query(f'total_id in @evs')['reco_energy'].to_numpy()  for evs in eventsF]
E_magic_marsF = [df_magic_mars.query(f'total_id in @evs')['reco_energy'].to_numpy() for evs in eventsF]
E_lst_mcpF    = [df_lst_mcp.query(f'total_id in @evs')['reco_energy'].to_numpy()    for evs in eventsF]
E_lst_lstF    = [df_lst_lst.query(f'total_id in @evs')['reco_energy'].to_numpy()    for evs in eventsF]

E_3T_mcpF_intensity = [df_3T_intensity.query(f'total_id in @evs')['reco_energy'].to_numpy() for evs in eventsF]
E_3T_mcpF_simple    = [df_3T_simple.query(f'total_id in @evs')['reco_energy'].to_numpy()    for evs in eventsF]

mask_red = f'total_id in @evs'
E_magic_mcpF_red  = [df_magic_mcp_red.query(mask_red)['reco_energy'].to_numpy()  for evs in eventsF]
E_3T_mcpF_red     = [df_3T_mcp_red.query(mask_red)['reco_energy'].to_numpy()  for evs in eventsF]

# computing energy ratios
r_LST_lst_mcp       = [E_lst_lstF[i] / E_lst_mcpF[i]    for i in range(len(gammas))]
r_3T_LST_mcp_lst    = [E_3T_mcpF[i]  / E_lst_lstF[i]    for i in range(len(gammas))]
r_LST_MAGIC_mcp_mcp = [E_lst_mcpF[i] / E_magic_mcpF[i]  for i in range(len(gammas))]

# mars with filtering
r_MAGIC_mcp_mars    = [E_magic_mcpF_red[i] / E_magic_marsF[i]    for i in range(len(gammas))]
r_3T_MAGIC_mcp_mars = [E_3T_mcpF_red[i]    / E_magic_marsF[i]    for i in range(len(gammas))]
r_MAGIC_mars_mcp    = [E_magic_marsF[i]    / E_magic_mcpF_red[i] for i in range(len(gammas))]
r_MAGIC_mars_mcp    = [E[~np.isnan(E)] for E in r_MAGIC_mars_mcp]

r_3T_var_intensity         = [E_3T_mcpF[i]           / E_3T_mcpF_intensity[i] for i in range(len(gammas))]
r_3T_var_simple            = [E_3T_mcpF[i]           / E_3T_mcpF_simple[i]    for i in range(len(gammas))]
r_3T_intensity_simple      = [E_3T_mcpF_intensity[i] / E_3T_mcpF_simple[i]    for i in range(len(gammas))]
r_3T_LST_mcp_intensity_lst = [E_3T_mcpF_intensity[i] / E_lst_lstF[i]          for i in range(len(gammas))]
r_3T_LST_mcp_simple_lst    = [E_3T_mcpF_simple[i]    / E_lst_lstF[i]          for i in range(len(gammas))]
```

## Energy distributions for all data types

```python
# binning
bins = np.logspace(-2.4, 2.7, 120)

fig, ax = plt.subplots(figsize=(10,6))
ax.set_title(f'Energy reconstruction {len(E_3T_mcp[0])} events')

ax.hist(E_3T_mcp[0], bins, color='c', alpha=1, histtype='stepfilled', label='stereo-3T')
ax.hist(E_lst_lst[0], bins, histtype='step',alpha=1, lw=3, color='deeppink', label='LST-lstchain')
ax.hist(E_lst_mcp[0], bins, histtype='step', lw=3, color='gold', label='LST-mcp')
ax.hist(E_magic_mcp[0], bins, alpha=1, color='darkblue', histtype='step', lw=3, label='MAGIC-mcp')
ax.hist(E_magic_mars[0], bins, alpha=1, color='crimson', histtype='step', lw=3, label='MAGIC-MARS')

ax.set_xlabel('Energy [TeV]')
ax.set_ylabel('counts')
ax.legend(loc=8)
plt.loglog()
plt.show()
```

## Comparison of `stereo-reco` between telescopes using `magic-cta-pipe`

```python
df_total_T1 = df_total.query(f'tel_id == 1')
df_total_T2 = df_total.query(f'tel_id == 2')
df_total_T3 = df_total.query(f'tel_id == 3')

# now we can filter in gamma but with the threshold given by 3-T stereo analysis with magic-mcp-pipe
dfs_3T  = [df_3T_mcp.query(f'gammaness > {g}', inplace=False) for g in gammas]
eventsF = [df['total_id'].to_numpy() for df in dfs_3T]

# filtering same events that the 3Tel stereo filtered in gammaness
E_3T  = E_3T_mcp.copy()
E_LST = [df_total_T1.query(f'total_id in @evs')['reco_energy'].to_numpy()  for evs in eventsF]
E_M1  = [df_total_T2.query(f'total_id in @evs')['reco_energy'].to_numpy()  for evs in eventsF]
E_M2  = [df_total_T3.query(f'total_id in @evs')['reco_energy'].to_numpy()  for evs in eventsF]

# computing energy ratios
r_3T_LST = [E_3T[i]  / E_LST[i] for i in range(len(gammas))]
r_3T_M1  = [E_3T[i]  / E_M1[i]  for i in range(len(gammas))]
r_3T_M2  = [E_3T[i]  / E_M2[i]  for i in range(len(gammas))]
r_M1_M2  = [E_M1[i]  / E_M2[i]  for i in range(len(gammas))]
r_LST_M1 = [E_LST[i] / E_M1[i]  for i in range(len(gammas))]
r_LST_M2 = [E_LST[i] / E_M2[i]  for i in range(len(gammas))]
```

```python
# gammaness index
gi = 5

# --- energy comparison --- #
bins = np.logspace(-2, 1.5, 120)
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(9.5,8.8), sharex=True, sharey=True)
fig.suptitle(f'stereo-reco for each telescope E-reconstruction $\gamma\geq {gammas[gi]}$ (mcp)')

ax1.hist2d(E_3T[gi], E_LST[gi],bins=bins, cmap='inferno')
ax2.hist2d(E_3T[gi],E_M1[gi],bins=bins, cmap='inferno')
ax3.hist2d(E_LST[gi], E_M1[gi],bins=bins, cmap='inferno')
ax4.hist2d(E_M1[gi], E_M2[gi],bins=bins, cmap='inferno')

ax1.set_title(f'3-Tel & LST-1')
ax2.set_title(f'3-Tel & M1')
ax3.set_title(f'LST-1 & M1')
ax4.set_title(f'M1 & M2')

ax1.set_xlabel('$E\ $(3-Tel)')
ax1.set_ylabel('$E\ $(LST-1)')
ax2.set_xlabel('$E\ $(3-Tel)')
ax2.set_ylabel('$E\ $(M1)')
ax3.set_xlabel('$E\ $(LST-1)')
ax3.set_ylabel('$E\ $(M1)')
ax4.set_xlabel('$E\ $(M1)')
ax4.set_ylabel('$E\ $(M2)')

for ax in [ax1, ax2, ax3, ax4]:
    ax.loglog()
    ax.plot([min(bins),max(bins)], [min(bins),max(bins)], lw=1 ,ls='--', c='w')
fig.tight_layout()
plt.show()


print(f'\nResults of different telescopes:\n')
m, s = np.mean(r_3T_LST[gi]), np.std(r_3T_LST[gi])
print(f'--> 3T/LST mean={m:.3f}, stdv={s:.3f}, |1-m|={abs(1-m):.3f}')
m, s = np.mean(r_3T_M1[gi]), np.std(r_3T_M1[gi])
print(f'--> 3T/M1  mean={m:.3f}, stdv={s:.3f}, |1-m|={abs(1-m):.3f}')
m, s = np.mean(r_LST_M1[gi]), np.std(r_LST_M1[gi])
print(f'--> LST/M1 mean={m:.3f}, stdv={s:.3f}, |1-m|={abs(1-m):.3f}')
m, s = np.mean(r_M1_M2[gi]), np.std(r_M1_M2[gi])
print(f'--> M1/M2  mean={m:.3f}, stdv={s:.3f}, |1-m|={abs(1-m):.3f}')

# --- energy ratios --- #
bins = np.linspace(0,2.5,100)

fig, ax = plt.subplots(figsize=(10.5,5))
ax.set_title(f'$\gamma\geq {gammas[gi]}$')

ax.hist(r_3T_LST[gi], bins=bins, label='3T/LST', histtype='stepfilled', color='c')
ax.hist(r_3T_M1[gi], bins=bins, label='3T/M1', histtype='step', lw=2, color='deeppink')
ax.hist(r_LST_M1[gi], bins=bins, label='LST/M1', histtype='step', lw=2, color='darkblue')
ax.hist(r_M1_M2[gi], bins=bins, label='M1/M2', histtype='step', lw=2, color='gold')

ax.axvline(1, color='k', lw=1, linestyle='--')

ax.legend()
ax.set_xlabel('$E_i/E_j$')
ax.set_ylabel('counts')
plt.show()
```

## Comparison between `weigth_types`

```python
# gammaness index
gi = 5
# --- energy ratios --- #
bins = np.linspace(0.6, 1.4, 200)

print(f'\nResults of different weigth methods:\n')
m, s = np.mean(r_3T_var_intensity[gi]), np.std(r_3T_var_intensity[gi])
print(f'--> "variance"/"intensity" mean={m:.3f}, stdv={s:.3f}, |1-m|={abs(1-m):.3f}')
m, s = np.mean(r_3T_var_simple[gi]), np.std(r_3T_var_simple[gi])
print(f'--> "variance"/"simple"    mean={m:.3f}, stdv={s:.3f}, |1-m|={abs(1-m):.3f}')
m, s = np.mean(r_3T_intensity_simple[gi]), np.std(r_3T_intensity_simple[gi])
print(f'--> "intensity"/"simple"   mean={m:.3f}, stdv={s:.3f}, |1-m|={abs(1-m):.3f}')

fig, ax = plt.subplots(figsize=(10.5,5))
ax.set_title(f'$\gamma\geq {gammas[gi]}$')

ax.hist(r_3T_var_intensity[gi], bins=bins, label='"variance"/"intensity"', histtype='stepfilled', lw=2, color='c')
ax.hist(r_3T_var_simple[gi], bins=bins, label='"variance"/"simple"', histtype='step', lw=2, color='deeppink')
ax.hist(r_3T_intensity_simple[gi], bins=bins, label='"intensity"/"simple"', histtype='step', lw=2, color='gold')

ax.axvline(1, color='k', lw=1, linestyle='--')

ax.legend()
ax.set_xlabel('$E_i/E_j$')
ax.set_ylabel('counts')
plt.show()
```
