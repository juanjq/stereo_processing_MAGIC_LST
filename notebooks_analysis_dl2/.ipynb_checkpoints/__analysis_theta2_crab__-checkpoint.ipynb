## Reading all dataframes

```python
# reading all data already produced
df_total      = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_total}.h5'))
df_3T_mcp     = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_3T}.h5'))
df_lst_mcp    = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_lst_mcp}.h5'))
df_magic_mcp  = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_stereo_magic}.h5'))
df_lst_lst    = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_lst_lst}.h5'))
df_magic_mars = pd.read_hdf(os.path.join(dir_stereo_reco, f'{name_dl2_magic}.h5'))
df_magic_mars.dropna(inplace=True)
```

## Extractting the values from the dataframes

```python
# energy of each dataframe filtered in gammaness
E_3T_mcp     = [df_3T_mcp['reco_energy'][df_3T_mcp['gammaness']>=g].to_numpy() for g in gammas]
E_magic_mcp  = [df_magic_mcp['reco_energy'][df_magic_mcp['gammaness']>=g].to_numpy() for g in gammas]
E_magic_mars = [df_magic_mars['reco_energy'][df_magic_mars['gammaness']>=g].to_numpy() for g in gammas]
E_lst_mcp    = [df_lst_mcp['reco_energy'][df_lst_mcp['gammaness']>=g].to_numpy() for g in gammas]
E_lst_lst    = [df_lst_lst['reco_energy'][df_lst_lst['gammaness']>=g].to_numpy() for g in gammas]

# thetas of each dataframe filtered in gammaness
Ton_3T_mcp     = [df_3T_mcp['theta2_on'][df_3T_mcp['gammaness']>=g].to_numpy() for g in gammas]
Ton_magic_mcp  = [df_magic_mcp['theta2_on'][df_magic_mcp['gammaness']>=g].to_numpy() for g in gammas]
Ton_magic_mars = [df_magic_mars['theta2_on'][df_magic_mars['gammaness']>=g].to_numpy() for g in gammas]
Ton_lst_mcp    = [df_lst_mcp['theta2_on'][df_lst_mcp['gammaness']>=g].to_numpy() for g in gammas]
Ton_lst_lst    = [df_lst_lst['theta2_on'][df_lst_lst['gammaness']>=g].to_numpy() for g in gammas]

Toff_3T_mcp     = [df_3T_mcp['theta2_off'][df_3T_mcp['gammaness']>=g].to_numpy() for g in gammas]
Toff_magic_mcp  = [df_magic_mcp['theta2_off'][df_magic_mcp['gammaness']>=g].to_numpy() for g in gammas]
Toff_magic_mars = [df_magic_mars['theta2_off'][df_magic_mars['gammaness']>=g].to_numpy() for g in gammas]
Toff_lst_mcp    = [df_lst_mcp['theta2_off'][df_lst_mcp['gammaness']>=g].to_numpy() for g in gammas]
Toff_lst_lst    = [df_lst_lst['theta2_off'][df_lst_lst['gammaness']>=g].to_numpy() for g in gammas]
```

## Energy distributions for all data types

```python
gamma_array = np.linspace(0, 1, 200)

time_diffs = np.diff(df_3T_mcp['timestamp'].to_numpy())
obs_time   = time_diffs[time_diffs < 1].sum() * u.s
print(f'The total observation time is: {obs_time.to(u.h):.2f}')

print(f'\nGlobal theta2 cut: {cut_value_theta2} [deg2]\n')

# Calculate the numbers ON, OFF and excess events
dfs      = [df_3T_mcp, df_lst_lst, df_lst_mcp, df_magic_mcp, df_magic_mars]
dfs      = [df[['gammaness', 'theta2_on', 'theta2_off', 'reco_energy']] for df in dfs]
label_df = ['3T-mcp', 'LST-lst', 'LST-mcp', 'MAGIC-mcp', 'MAGIC-mars']
cols     = ['darkblue', 'darkviolet', 'crimson', 'gold', 'pink']

significance = [[] for _ in dfs]
prop_gamma   = [[] for _ in dfs]
gamma_rate   = [[] for _ in dfs]
off_rate     = [[] for _ in dfs]

tmp_dfs = dfs.copy()

for g in range(len(gamma_array)):
    
    if g % 20 == 0:
        print(f'Computing {g / len(gamma_array) * 100:.0f}%...')
        
    for i in range(len(dfs)):
        
        tmp_dfs[i] = tmp_dfs[i].query(f'gammaness >= {gamma_array[g]}', inplace=False)

        n_on  = sum(tmp_dfs[i]['theta2_on' ] < cut_value_theta2)
        n_off = sum(tmp_dfs[i]['theta2_off'] < cut_value_theta2)

        significance[i].append(WStatCountsStatistic(n_on, n_off, alpha=1).sqrt_ts)
        
        gamma_rate[i].append((n_on-n_off)/obs_time.to(u.min).value)
        off_rate[i].append(n_off/obs_time.to(u.min).value)
        
        if n_off !=0:
            prop_gamma[i].append(n_on/n_off)
        else:
            prop_gamma[i].append(0)
```

```python
# calculate the optimal gammas
def neg_WStat(gammaness, i):
    
    df = dfs[i].query(f'gammaness >= {gammaness}', inplace=False)[['gammaness', 'theta2_on', 'theta2_off']]
    
    n_on  = sum(df['theta2_on' ] < cut_value_theta2)
    n_off = sum(df['theta2_off'] < cut_value_theta2)
        
    return -1 * WStatCountsStatistic(n_on, n_off, alpha=1).sqrt_ts

opt_gamma = []
for i in range(len(significance)):
    opt_gamma.append(minimize(neg_WStat, x0=[0.7], bounds=[[0,1]], args=(i), method='Nelder-Mead').x[0])
    
```

```python
print(f'The maxiumum significances are given by:')
for i in range(len(significance)):
    maxgamma, maxsig = gamma_array[significance[i].index(max(significance[i]))], max(significance[i])
    nevs  = len(dfs[i][dfs[i]['gammaness'] >= maxgamma])
    nevsp = nevs / len(dfs[i]) * 100
    print(f'--> {label_df[i]}\t gamma = {opt_gamma[i]:.3f},  sigma = {maxsig:.1f},  nevs = {nevs}, {nevsp:.1f}%')

fig, ax = plt.subplots(figsize=(10,6))
ax.set_title(f'Significance and gammaness cuts')

for i in range(len(significance)-1):
    ax.plot(gamma_array, significance[i], label=label_df[i], c=cols[i])
    ax.axvline(opt_gamma[i], color=cols[i], lw=2, linestyle='--')

ax.set_xlabel('$\gamma$-cut')
ax.set_ylabel('significance-$\sigma$')
ax.legend(loc=8)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7.5,5))
ax.set_title(f'ON/OFF ratio and gammaness cuts')

for i in range(len(significance)-1):
    ax.plot(gamma_array, prop_gamma[i], label=label_df[i], c=cols[i])

ax.set_xlabel('$\gamma$-cut')
ax.set_ylabel('n_on/n_off')
ax.legend(loc=2)
ax.set_xlim(0.8, 1.01)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.set_title(f'rates and gammaness cuts')

for i in range(len(significance)-1):
    ax.plot(gamma_array, gamma_rate[i], label=label_df[i], c=cols[i], linestyle='-', lw=3)
    ax.plot(gamma_array, off_rate[i], c=cols[i], linestyle='--', lw=3)

ax.plot([], [], c='gray', label='gamma-rate', ls='-', lw=3)
ax.plot([], [], c='gray', label='off-rate', ls='--', lw=3)

ax.set_xlabel('$\gamma$-cut')
ax.set_ylabel('rate [ev/min]')
ax.legend(loc=1)
plt.show()
```

## Angular resolution

```python
energy_bins = np.logspace(-2, 2, 10)

energy_bins_center = (energy_bins[:-1] + energy_bins[1:]) / 2

energy_bins_width = [
    energy_bins[1:] - energy_bins_center,
    energy_bins_center - energy_bins[:-1],
]


angres = []
for df in dfs[:-1]:

    # get event theta value
    theta2vec= df["theta2_on"]

    thetavec=u.Quantity(np.sqrt(theta2vec), u.deg)
    # add theta value to dataframe and get a QTable from it
    df['theta']=thetavec


    df_table = QTable.from_pandas(df)
    df_table[:5]
    # Apply global theta cut to events
    mask = df_table['theta'].value<np.repeat(np.sqrt(0.04),(len(df_table)))
    df_table=df_table[mask]
    df_table['reco_energy']=u.Quantity(df_table['reco_energy'],u.TeV)
    # angular resolution
    angres_table_Crab = angular_resolution(df_table, u.Quantity(energy_bins,u.TeV), energy_type="reco"
    )

    angres_Crab = angres_table_Crab["angular_resolution"].value
    
    angres.append(angres_Crab)
# print(f"Angular resolution:\n{angres_Crab} deg")
# print(f"Energy bin center:\n{energy_bins_center} TeV")
```

```python
fig, ax = plt.subplots(figsize=(10,7))

plt.title(f"angular resolution")
plt.ylabel("Angular resolution (68% cont.) [deg]")
plt.xlabel("E [TeV]")
plt.semilogx()
plt.grid()

for i in range(len(angres)):
    ax.errorbar(x=energy_bins_center, y=angres[i], xerr=energy_bins_width,label=label_df[i], marker="o", 
                color=cols[i], )

ax.legend()
plt.show()
```

```python
def get_68_containment(dist, bin_edges):
    
    distribution = np.array(dist) / sum(dist)

    bin_widths = np.diff(bin_edges)
    bin_centers = bin_edges[:-1] + bin_widths / 2

    # calculate the cumulative sum of the sorted array
    cumulative_sum = np.cumsum(distribution)

    # find the indices where the cumulative sum is within the 68% range
    for i in range(len(cumulative_sum)):
        if cumulative_sum[i] <= 0.68:
            lower_idx = i
        if cumulative_sum[i] >= 0.68:
            upper_idx = i
            break

    # interpolate between the two indices to find the exact location of the 68%
    resto = 0.68 - cumulative_sum[lower_idx]
    Dc = cumulative_sum[upper_idx] - cumulative_sum[lower_idx]
    Dd = bin_edges[upper_idx+1] - bin_edges[lower_idx+1]
    
    location = bin_edges[lower_idx+1] + Dd * (resto / Dc)

    return location
```

```python
nbins = 11

df = df_lst_mcp.query(f'gammaness > 0.7')

E  = df['reco_energy'].to_numpy()
Ebins = np.logspace(np.log10(min(E)),1, nbins)
Ebins_c =  (Ebins[:-1] + Ebins[1:]) / 2

Gbins = np.linspace(0, 4, 200)
Gbins_c =  (Gbins[:-1] + Gbins[1:]) / 2

t_on, t_off = [], []
for i in range(len(Ebins)-1):
    mask = f'reco_energy >= {Ebins[i]} and reco_energy <= {Ebins[i+1]}'
    t_on.append(df.query(mask)['theta2_on'].to_numpy())
    t_off.append(df.query(mask)['theta2_off'].to_numpy())
    
# now we calculate the distribution on - off
dist = []
containment = []
for i in range(len(t_on)):
    
    h_on, edeges  = np.histogram(t_on[i],  bins=Gbins)
    h_off, edeges = np.histogram(t_off[i], bins=Gbins)
    
    tmp_dist = h_on - h_off
    
    # negative --> 0
    tmp_dist = [max(tmp, 0) for tmp in tmp_dist]
    
    dist.append(tmp_dist)
    
    containment.append(get_68_containment(tmp_dist, Gbins))
```

```python
containmentLST = containment.copy()
```

```python
nrows = (len(Ebins) - 1) // 2
fig, axe = plt.subplots(nrows, 2, figsize=(10, 4*nrows), sharex=True, sharey='row')

axes = [axe[i][j] for i in range(len(axe)) for j in range(len(axe[i])) ]

for i, ax in zip(range(len(axes)), axes):
    
    ax.hist(t_on[i], bins=Gbins, histtype='step', color='darkblue')
    ax.hist(t_off[i], bins=Gbins, histtype='step', color='c')
    
    ax.set_title(f'{Ebins[i]:.0e} > E > {Ebins[i+1]:.0e} [TeV]')
    
    ax.axvline(containment[i], color='k', linestyle='--', lw=1.5, )
    
    if i%2 == 0:
        ax.set_ylabel('counts')

axes[-1].set_xlabel(f'$\\theta^2$ [deg]')
axes[-2].set_xlabel(f'$\\theta^2$ [deg]')

plt.show()
```
